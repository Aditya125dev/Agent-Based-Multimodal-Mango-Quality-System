```python
# ==========================================================================================================
# HARUMANIS MANGO GRADING SYSTEM
# ==========================================================================================================
# Purpose: Automated mango quality assessment using computer vision and machine learning
# Features:
#   - Multi-view image analysis (front + back)
#   - Weight-based grading
#   - Ripeness and grade classification
#   - Market price estimation (RM & INR)
#   - Comprehensive validation (blur, brightness, authenticity)
# Model: Gradient Boosting Classifier with K-Fold Cross-Validation
# Standard: MS 885:2012 (Malaysian Standard for Harumanis Mango)
# ==========================================================================================================

import os
import cv2
import numpy as np
import pandas as pd
import warnings

from skimage.feature import graycomatrix, graycoprops
from skimage.metrics import structural_similarity as ssim

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import GradientBoostingClassifier

warnings.filterwarnings("ignore")

print("="*90)
print("HARUMANIS MANGO GRADING SYSTEM - INTELLIGENT QUALITY ASSESSMENT".center(90))
print("="*90)

# ==========================================================================================================
# CONFIGURATION - FILE PATHS
# ==========================================================================================================

IMAGE_DIR = "/content/images"              # Directory containing mango images
EXCEL_PATH = "/content/Harumanis_mango_weight_grade.xlsx"  # Dataset with labels

# ==========================================================================================================
# FUNCTION 1: MANGO SEGMENTATION
# ==========================================================================================================
# Purpose: Remove background from image to isolate mango fruit
# Method: HSV color space thresholding + morphological operations
# Input: BGR image
# Output: Segmented mango image + binary mask
# ==========================================================================================================

def segment_mango(img):
    """
    Segment mango from background using HSV color thresholding

    Args:
        img: Input BGR image

    Returns:
        mango: Segmented mango (background removed)
        mask: Binary mask of mango region
    """
    # Convert BGR to HSV color space (better for color-based segmentation)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Define HSV range for mango (green to yellow hues)
    lower = np.array([15, 30, 30])    # Lower bound: H=15, S=30, V=30
    upper = np.array([95, 255, 255])  # Upper bound: H=95, S=255, V=255

    # Create binary mask
    mask = cv2.inRange(hsv, lower, upper)

    # Morphological closing to fill gaps
    kernel = np.ones((7, 7), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    # Apply mask to original image
    mango = cv2.bitwise_and(img, img, mask=mask)

    return mango, mask


# ==========================================================================================================
# FUNCTION 2: BLUR DETECTION
# ==========================================================================================================
# Purpose: Detect if image is too blurry for reliable analysis
# Method: Laplacian variance (measures focus/sharpness)
# Threshold: < 35 indicates blurry image
# ==========================================================================================================

def check_blur(img):
    """
    Detect blurry images using Laplacian variance

    Args:
        img: Input BGR image

    Returns:
        is_blurry: Boolean (True if blurry)
        blur_score: Numeric score (higher = sharper)
    """
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Calculate Laplacian variance (edge detection)
    blur_score = cv2.Laplacian(gray, cv2.CV_64F).var()

    # Threshold: < 35 is considered blurry
    is_blurry = blur_score < 35

    return is_blurry, blur_score


# ==========================================================================================================
# FUNCTION 3: BRIGHTNESS VALIDATION (NEW)
# ==========================================================================================================
# Purpose: Check if image has proper lighting for analysis
# Method: Calculate mean brightness in HSV V-channel
# Thresholds: < 40 (too dark), > 220 (too bright), 40-220 (good)
# ==========================================================================================================

def check_brightness(img):
    """
    Validate image brightness/lighting conditions

    Args:
        img: Input BGR image

    Returns:
        status: "too_dark", "too_bright", or "ok"
        brightness: Numeric brightness value (0-255)
    """
    # Convert to HSV (V-channel = brightness)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Calculate mean brightness
    brightness = hsv[:, :, 2].mean()

    # Determine status
    if brightness < 40:
        return "too_dark", brightness
    elif brightness > 220:
        return "too_bright", brightness
    else:
        return "ok", brightness


# ==========================================================================================================
# FUNCTION 4: EXACT DUPLICATE DETECTION
# ==========================================================================================================
# Purpose: Detect if front and back images are identical (same photo uploaded twice)
# Method: Structural Similarity Index (SSIM)
# Threshold: > 0.95 indicates duplicate
# ==========================================================================================================

def check_exact_duplicate(img1, img2):
    """
    Check if two images are exact duplicates

    Args:
        img1, img2: Input BGR images

    Returns:
        is_duplicate: Boolean (True if same image)
        ssim_score: Similarity score (0-1, higher = more similar)
    """
    # Convert to grayscale
    gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

    # Resize to match if needed
    if gray1.shape != gray2.shape:
        gray2 = cv2.resize(gray2, (gray1.shape[1], gray1.shape[0]))

    # Calculate SSIM (Structural Similarity Index)
    ssim_score, _ = ssim(gray1, gray2, full=True)

    # Threshold: > 0.95 means images are duplicates
    is_duplicate = ssim_score > 0.95

    return is_duplicate, ssim_score


# ==========================================================================================================
# FUNCTION 5: SAME MANGO VERIFICATION (MULTI-CHECK)
# ==========================================================================================================
# Purpose: Verify that front and back images are from THE SAME mango (not different fruits)
# Method: Triple validation using:
#   1. Color histogram correlation (HSV space)
#   2. Structural similarity (SSIM)
#   3. ORB feature matching (keypoint descriptors)
# Decision: ALL 3 checks must pass to confirm same mango
# ==========================================================================================================

def check_same_mango(img1, img2):
    """
    Verify images are from the same mango using multi-check validation

    Args:
        img1, img2: Input BGR images

    Returns:
        is_same: Boolean (True if same mango)
        details: Dictionary with individual check scores
    """
    # Segment mangoes to remove background
    mango1, _ = segment_mango(img1)
    mango2, _ = segment_mango(img2)

    # -------------------------
    # CHECK 1: Color Histogram Correlation
    # -------------------------
    # Compare color distributions in HSV space
    hsv1 = cv2.cvtColor(mango1, cv2.COLOR_BGR2HSV)
    hsv2 = cv2.cvtColor(mango2, cv2.COLOR_BGR2HSV)

    # Calculate 2D histograms (Hue + Saturation)
    hist1 = cv2.calcHist([hsv1], [0, 1], None, [50, 60], [0, 180, 0, 256])
    hist2 = cv2.calcHist([hsv2], [0, 1], None, [50, 60], [0, 180, 0, 256])

    # Normalize histograms
    cv2.normalize(hist1, hist1)
    cv2.normalize(hist2, hist2)

    # Calculate correlation (1 = identical, 0 = completely different)
    hist_sim = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)

    # -------------------------
    # CHECK 2: Structural Similarity (SSIM)
    # -------------------------
    # Compare overall structure/shape
    g1 = cv2.cvtColor(mango1, cv2.COLOR_BGR2GRAY)
    g2 = cv2.cvtColor(mango2, cv2.COLOR_BGR2GRAY)
    g2 = cv2.resize(g2, (g1.shape[1], g1.shape[0]))

    ssim_score, _ = ssim(g1, g2, full=True)

    # -------------------------
    # CHECK 3: ORB Feature Matching
    # -------------------------
    # Compare local features (keypoints + descriptors)
    orb = cv2.ORB_create()  # ORB = Oriented FAST and Rotated BRIEF

    # Detect keypoints and compute descriptors
    kp1, des1 = orb.detectAndCompute(g1, None)
    kp2, des2 = orb.detectAndCompute(g2, None)

    matches = 0
    if des1 is not None and des2 is not None:
        # Brute-force matcher with Hamming distance
        bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
        matches = len(bf.match(des1, des2))

    # -------------------------
    # FINAL DECISION
    # -------------------------
    # All 3 checks must pass:
    #   - Histogram correlation >= 0.85
    #   - SSIM >= 0.75
    #   - ORB matches >= 8
    is_same = (hist_sim >= 0.85 and ssim_score >= 0.75 and matches >= 8)

    details = {
        'hist_sim': hist_sim,
        'ssim': ssim_score,
        'orb_matches': matches
    }

    return is_same, details


# ==========================================================================================================
# FUNCTION 6: WEIGHT VALIDATION
# ==========================================================================================================
# Purpose: Validate mango weight is within reasonable range
# Range: 300-700 grams (typical for Harumanis variety)
# ==========================================================================================================

def check_weight(weight):
    """
    Validate weight is within acceptable range for Harumanis mangoes

    Args:
        weight: Weight in grams

    Returns:
        is_valid: Boolean (True if valid)
    """
    return 300 <= weight <= 700


# ==========================================================================================================
# FUNCTION 7: IMAGE FEATURE EXTRACTION
# ==========================================================================================================
# Purpose: Extract numerical features from mango images for ML model
# Features: 37 total
#   - RGB statistics (6): mean and std for each channel
#   - HSV statistics (6): mean and std for H, S, V
#   - Hue histogram (16): distribution of hue values
#   - Texture features (5): contrast, homogeneity, energy, correlation, dissimilarity
#   - Color ratios (4): yellow ratio, green ratio, saturation stats, value stats
# ==========================================================================================================

def extract_image_features(img):
    """
    Extract 37 numerical features from mango image

    Features extracted:
        - Color statistics (RGB + HSV)
        - Hue histogram (ripeness indicator)
        - Texture features (GLCM-based)

    Args:
        img: Input BGR image

    Returns:
        features: 1D numpy array of 37 features
    """
    # Resize to standard size for consistency
    img = cv2.resize(img, (224, 224))

    # -------------------------
    # RGB Color Statistics
    # -------------------------
    rgb_mean = img.mean(axis=(0, 1))  # Mean of each channel
    rgb_std = img.std(axis=(0, 1))    # Standard deviation

    # -------------------------
    # HSV Color Statistics
    # -------------------------
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0, 1))
    hsv_std = hsv.std(axis=(0, 1))

    # -------------------------
    # Hue Histogram (Ripeness Indicator)
    # -------------------------
    hue = hsv[:, :, 0]
    hue_hist, _ = np.histogram(hue, bins=16, range=(0, 180))
    hue_hist = hue_hist / (hue_hist.sum() + 1e-10)  # Normalize

    # -------------------------
    # Texture Features (GLCM)
    # -------------------------
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_r = (gray / 4).astype(np.uint8)  # Reduce to 64 levels

    # Gray-Level Co-occurrence Matrix
    glcm = graycomatrix(
        gray_r,
        distances=[1],
        angles=[0, np.pi/4, np.pi/2, 3*np.pi/4],
        levels=64,
        symmetric=True,
        normed=True
    )

    # Extract texture properties
    texture = [
        graycoprops(glcm, 'contrast').mean(),
        graycoprops(glcm, 'homogeneity').mean(),
        graycoprops(glcm, 'energy').mean(),
        graycoprops(glcm, 'correlation').mean(),
        graycoprops(glcm, 'dissimilarity').mean()
    ]

    # Combine all features
    return np.hstack([rgb_mean, rgb_std, hsv_mean, hsv_std, hue_hist, texture])


# ==========================================================================================================
# STEP 1: DATA LOADING
# ==========================================================================================================
# Load dataset from Excel file containing:
#   - Fruit No (image filename)
#   - Mass(kg) (weight)
#   - Color_K-Yellow_P_Green (ripeness label)
#   - Fruit Grade (quality grade: P/1/2)
# ==========================================================================================================

print("\n" + "="*90)
print("[STEP 1/6] LOADING DATASET".center(90))
print("="*90)

df = pd.read_excel(EXCEL_PATH)

# Prepare data
df["Fruit Grade"] = df["Fruit Grade"].astype(str)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

print(f"✓ Loaded {len(df)} records from Excel")
print(f"✓ Ripeness distribution:\n{df['Color_K-Yellow_P_Green'].value_counts()}")
print(f"✓ Grade distribution:\n{df['Fruit Grade'].value_counts()}")


# ==========================================================================================================
# STEP 2: FEATURE EXTRACTION & STAGE EXCEL GENERATION
# ==========================================================================================================
# Process each mango (requires 2 images: front + back)
# Generate 4 Excel files documenting feature extraction stages:
#   Stage 1: Front view features (37 features)
#   Stage 2: Back view features (37 features)
#   Stage 3: Combined features (74 features = front + back)
#   Stage 4: Fused features (37 averaged + 1 weight = 38 features)
# ==========================================================================================================

print("\n" + "="*90)
print("[STEP 2/6] FEATURE EXTRACTION & STAGE DOCUMENTATION".center(90))
print("="*90)

# Storage for different feature stages
stage1_front = []    # Front view features
stage2_back = []     # Back view features
stage3_combined = [] # Front + back concatenated
stage4_fused = []    # Averaged + weight

X_img = []      # Final image features (averaged)
X_weight = []   # Weight features
y_rip = []      # Ripeness labels
y_grd = []      # Grade labels
groups = []     # Fruit IDs for grouping

loaded = 0
skipped = 0

# Process each fruit
for fid, g in df.groupby("Fruit_ID"):
    # Each fruit should have exactly 2 images (front + back)
    if len(g) != 2:
        skipped += 1
        continue

    r1, r2 = g.iloc[0], g.iloc[1]

    # Load images
    img1 = cv2.imread(os.path.join(IMAGE_DIR, r1["Fruit No"]))
    img2 = cv2.imread(os.path.join(IMAGE_DIR, r2["Fruit No"]))

    if img1 is None or img2 is None:
        skipped += 1
        continue

    # Extract features from both views
    f1 = extract_image_features(img1)  # Front: 37 features
    f2 = extract_image_features(img2)  # Back: 37 features

    # Average features from both views
    feat_avg = (f1 + f2) / 2  # Fused: 37 features

    # Get weight and labels
    weight = r1["Mass(kg)"]
    ripeness = r1["Color_K-Yellow_P_Green"]
    grade = r1["Fruit Grade"]

    # -------------------------
    # STAGE 1: Front View
    # -------------------------
    stage1_row = {
        'Fruit_ID': fid,
        'Image_Name': r1["Fruit No"],
        'Weight_kg': weight,
        'Ripeness': ripeness,
        'Grade': grade
    }
    # Add 37 features with names
    feature_names = [
        'rgb_b_mean', 'rgb_g_mean', 'rgb_r_mean',
        'rgb_b_std', 'rgb_g_std', 'rgb_r_std',
        'hsv_h_mean', 'hsv_s_mean', 'hsv_v_mean',
        'hsv_h_std', 'hsv_s_std', 'hsv_v_std'
    ] + [f'hue_bin_{i}' for i in range(16)] + [
        'contrast', 'homogeneity', 'energy', 'correlation', 'dissimilarity'
    ]

    for i, fname in enumerate(feature_names):
        stage1_row[f'front_{fname}'] = f1[i]
    stage1_front.append(stage1_row)

    # -------------------------
    # STAGE 2: Back View
    # -------------------------
    stage2_row = {
        'Fruit_ID': fid,
        'Image_Name': r2["Fruit No"],
        'Weight_kg': weight,
        'Ripeness': ripeness,
        'Grade': grade
    }
    for i, fname in enumerate(feature_names):
        stage2_row[f'back_{fname}'] = f2[i]
    stage2_back.append(stage2_row)

    # -------------------------
    # STAGE 3: Combined (Front + Back)
    # -------------------------
    stage3_row = {
        'Fruit_ID': fid,
        'Front_Image': r1["Fruit No"],
        'Back_Image': r2["Fruit No"],
        'Weight_kg': weight,
        'Ripeness': ripeness,
        'Grade': grade
    }
    for i, fname in enumerate(feature_names):
        stage3_row[f'front_{fname}'] = f1[i]
        stage3_row[f'back_{fname}'] = f2[i]
    stage3_combined.append(stage3_row)

    # -------------------------
    # STAGE 4: Fused (Averaged + Weight)
    # -------------------------
    stage4_row = {
        'Fruit_ID': fid,
        'Front_Image': r1["Fruit No"],
        'Back_Image': r2["Fruit No"],
        'Weight_kg': weight,
        'Ripeness': ripeness,
        'Grade': grade
    }
    for i, fname in enumerate(feature_names):
        stage4_row[f'fused_{fname}'] = feat_avg[i]
    stage4_row['physical_weight'] = weight
    stage4_fused.append(stage4_row)

    # Store for model training
    X_img.append(feat_avg)
    X_weight.append([weight])
    y_rip.append(ripeness)
    y_grd.append(grade)
    groups.append(fid)

    loaded += 1

print(f"✓ Processed {loaded} mangoes | Skipped {skipped}")

# -------------------------
# Save Stage Excel Files
# -------------------------
print("\nGenerating stage documentation Excel files...")

pd.DataFrame(stage1_front).to_excel("stage1_front_view_features.xlsx", index=False)
print("  ✓ Stage 1: stage1_front_view_features.xlsx (Front view - 37 features)")

pd.DataFrame(stage2_back).to_excel("stage2_back_view_features.xlsx", index=False)
print("  ✓ Stage 2: stage2_back_view_features.xlsx (Back view - 37 features)")

pd.DataFrame(stage3_combined).to_excel("stage3_combined_features.xlsx", index=False)
print("  ✓ Stage 3: stage3_combined_features.xlsx (Front + Back - 74 features)")

pd.DataFrame(stage4_fused).to_excel("stage4_fused_with_weight.xlsx", index=False)
print("  ✓ Stage 4: stage4_fused_with_weight.xlsx (Averaged + Weight - 38 features)")


# ==========================================================================================================
# STEP 3: DATA PREPARATION
# ==========================================================================================================
# Combine image features with weight for final feature vector
# Encode categorical labels (ripeness and grade) to numeric values
# ==========================================================================================================

print("\n" + "="*90)
print("[STEP 3/6] DATA PREPARATION".center(90))
print("="*90)

# Combine features: [37 image features] + [1 weight] = 38 features
X = np.hstack([np.array(X_img), np.array(X_weight)])
groups = np.array(groups)

print(f"✓ Final feature matrix shape: {X.shape}")

# -------------------------
# Label Encoding
# -------------------------
rip_enc = LabelEncoder()  # Ripeness encoder
grd_enc = LabelEncoder()  # Grade encoder

y_rip = rip_enc.fit_transform(y_rip)
y_grd = grd_enc.fit_transform(y_grd)

print(f"✓ Ripeness classes: {rip_enc.classes_}")
print(f"✓ Grade classes: {grd_enc.classes_}")


# ==========================================================================================================
# STEP 4: K-FOLD CROSS-VALIDATION
# ==========================================================================================================
# Purpose: Robust evaluation to prevent overfitting
# Method: 5-Fold Stratified Cross-Validation
#   - Data split into 5 folds
#   - Train on 4 folds, test on 1 fold
#   - Repeat 5 times, each fold used as test once
#   - Report mean ± std of metrics
# Anti-Overfitting Measures:
#   - Reduced tree depth (max_depth=3)
#   - Minimum samples per split (min_samples_split=10)
#   - Minimum samples per leaf (min_samples_leaf=5)
#   - Subsampling (subsample=0.8)
#   - Low learning rate (learning_rate=0.05)
# ==========================================================================================================

print("\n" + "="*90)
print("[STEP 4/6] K-FOLD CROSS-VALIDATION (ANTI-OVERFITTING)".center(90))
print("="*90)

n_splits = 5
kfold = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

cv_results_rip = []  # Ripeness results
cv_results_grd = []  # Grade results

print(f"\nRunning {n_splits}-Fold Cross-Validation...\n")

for fold, (train_idx, test_idx) in enumerate(kfold.split(X, y_rip), 1):
    # Split data
    X_train, X_test = X[train_idx], X[test_idx]
    yr_train, yr_test = y_rip[train_idx], y_rip[test_idx]
    yg_train, yg_test = y_grd[train_idx], y_grd[test_idx]

    # Feature scaling (fit on train, transform on test)
    scaler = RobustScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_test_s = scaler.transform(X_test)

    # Train Gradient Boosting models with anti-overfitting parameters
    gb_rip = GradientBoostingClassifier(
        n_estimators=50,           # Number of trees
        learning_rate=0.05,        # Low learning rate (more conservative)
        max_depth=3,               # Shallow trees
        min_samples_split=10,      # Minimum samples to split node
        min_samples_leaf=5,        # Minimum samples per leaf
        subsample=0.8,             # Use 80% of data per tree
        random_state=42
    )

    gb_grd = GradientBoostingClassifier(
        n_estimators=50,
        learning_rate=0.05,
        max_depth=3,
        min_samples_split=10,
        min_samples_leaf=5,
        subsample=0.8,
        random_state=42
    )

    # Train models
    gb_rip.fit(X_train_s, yr_train)
    gb_grd.fit(X_train_s, yg_train)

    # Predict
    yr_pred = gb_rip.predict(X_test_s)
    yg_pred = gb_grd.predict(X_test_s)

    # Calculate metrics
    cv_results_rip.append({
        'fold': fold,
        'acc': accuracy_score(yr_test, yr_pred),
        'prec': precision_score(yr_test, yr_pred, average='weighted', zero_division=0),
        'rec': recall_score(yr_test, yr_pred, average='weighted', zero_division=0),
        'f1': f1_score(yr_test, yr_pred, average='weighted', zero_division=0)
    })

    cv_results_grd.append({
        'fold': fold,
        'acc': accuracy_score(yg_test, yg_pred),
        'prec': precision_score(yg_test, yg_pred, average='weighted', zero_division=0),
        'rec': recall_score(yg_test, yg_pred, average='weighted', zero_division=0),
        'f1': f1_score(yg_test, yg_pred, average='weighted', zero_division=0)
    })

    print(f"  Fold {fold}/{n_splits}: Ripeness F1={cv_results_rip[-1]['f1']:.4f} | Grade F1={cv_results_grd[-1]['f1']:.4f}")

# -------------------------
# Calculate Average Metrics
# -------------------------
avg_rip = {k: np.mean([r[k] for r in cv_results_rip]) for k in ['acc', 'prec', 'rec', 'f1']}
avg_grd = {k: np.mean([r[k] for r in cv_results_grd]) for k in ['acc', 'prec', 'rec', 'f1']}

std_rip = {k: np.std([r[k] for r in cv_results_rip]) for k in ['acc', 'prec', 'rec', 'f1']}
std_grd = {k: np.std([r[k] for r in cv_results_grd]) for k in ['acc', 'prec', 'rec', 'f1']}

# -------------------------
# Display Results
# -------------------------
print("\n" + "="*90)
print("CROSS-VALIDATION RESULTS".center(90))
print("="*90)

print(f"\n{'RIPENESS CLASSIFICATION':^90}")
print("-"*90)
print(f"{'Metric':<20} {'Mean':>15} {'Std Dev':>15} {'Range':>30}")
print("-"*90)
print(f"{'Accuracy':<20} {avg_rip['acc']:>15.4f} {std_rip['acc']:>15.4f} {f'[{avg_rip['acc']-std_rip['acc']:.4f}, {avg_rip['acc']+std_rip['acc']:.4f}]':>30}")
print(f"{'Precision':<20} {avg_rip['prec']:>15.4f} {std_rip['prec']:>15.4f} {f'[{avg_rip['prec']-std_rip['prec']:.4f}, {avg_rip['prec']+std_rip['prec']:.4f}]':>30}")
print(f"{'Recall':<20} {avg_rip['rec']:>15.4f} {std_rip['rec']:>15.4f} {f'[{avg_rip['rec']-std_rip['rec']:.4f}, {avg_rip['rec']+std_rip['rec']:.4f}]':>30}")
print(f"{'F1-Score':<20} {avg_rip['f1']:>15.4f} {std_rip['f1']:>15.4f} {f'[{avg_rip['f1']-std_rip['f1']:.4f}, {avg_rip['f1']+std_rip['f1']:.4f}]':>30}")

print(f"\n{'GRADE CLASSIFICATION':^90}")
print("-"*90)
print(f"{'Metric':<20} {'Mean':>15} {'Std Dev':>15} {'Range':>30}")
print("-"*90)
print(f"{'Accuracy':<20} {avg_grd['acc']:>15.4f} {std_grd['acc']:>15.4f} {f'[{avg_grd['acc']-std_grd['acc']:.4f}, {avg_grd['acc']+std_grd['acc']:.4f}]':>30}")
print(f"{'Precision':<20} {avg_grd['prec']:>15.4f} {std_grd['prec']:>15.4f} {f'[{avg_grd['prec']-std_grd['prec']:.4f}, {avg_grd['prec']+std_grd['prec']:.4f}]':>30}")
print(f"{'Recall':<20} {avg_grd['rec']:>15.4f} {std_grd['rec']:>15.4f} {f'[{avg_grd['rec']-std_grd['rec']:.4f}, {avg_grd['rec']+std_grd['rec']:.4f}]':>30}")
print(f"{'F1-Score':<20} {avg_grd['f1']:>15.4f} {std_grd['f1']:>15.4f} {f'[{avg_grd['f1']-std_grd['f1']:.4f}, {avg_grd['f1']+std_grd['f1']:.4f}]':>30}")

print("="*90)


# ==========================================================================================================
# STEP 5: FINAL MODEL TRAINING
# ==========================================================================================================
# Train final models on ALL available data for deployment
# These models will be used for real-time prediction
# ==========================================================================================================

print("\n" + "="*90)
print("[STEP 5/6] TRAINING FINAL MODELS ON COMPLETE DATASET".center(90))
print("="*90)

# Scale all data
final_scaler = RobustScaler()
X_final = final_scaler.fit_transform(X)

# Train final models
final_gb_rip = GradientBoostingClassifier(
    n_estimators=50,
    learning_rate=0.05,
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    subsample=0.8,
    random_state=42
)

final_gb_grd = GradientBoostingClassifier(
    n_estimators=50,
    learning_rate=0.05,
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    subsample=0.8,
    random_state=42
)

print("\nTraining Ripeness model...", end=' ')
final_gb_rip.fit(X_final, y_rip)
print("✓")

print("Training Grade model...", end=' ')
final_gb_grd.fit(X_final, y_grd)
print("✓")

print("\n✓ Final models ready for deployment")


# ==========================================================================================================
# FUNCTION 8: PRICE CALCULATION (RM + INR)
# ==========================================================================================================
# Purpose: Estimate market price based on grade, ripeness, and weight
# Currency: Malaysian Ringgit (RM) and Indian Rupees (INR)
# Conversion Rate: 1 RM = 18.5 INR (approximate)
# Price Matrix (RM):
#   Premium (P) + Ripe (K): 40 RM
#   Premium (P) + Unripe (P): 35 RM
#   Grade 1 + Ripe (K): 28 RM
#   Grade 1 + Unripe (P): 24 RM
#   Grade 2 + Ripe (K): 18 RM
#   Grade 2 + Unripe (P): 15 RM
# Adjustments:
#   +5 RM if weight > 550g (large fruit)
#   -3 RM if weight < 350g (small fruit)
# ==========================================================================================================

def price_decision(grade, ripeness, weight):
    """
    Calculate market price in RM and INR

    Args:
        grade: Quality grade (P/1/2)
        ripeness: Ripeness stage (K=ripe, P=unripe)
        weight: Weight in grams

    Returns:
        price_rm: Price in Malaysian Ringgit
        price_inr: Price in Indian Rupees
    """
    # Base price matrix (RM)
    if grade == "P" and ripeness == "K":
        price_rm = 40
    elif grade == "P" and ripeness == "P":
        price_rm = 35
    elif grade == "1" and ripeness == "K":
        price_rm = 28
    elif grade == "1" and ripeness == "P":
        price_rm = 24
    elif grade == "2" and ripeness == "K":
        price_rm = 18
    else:  # Grade 2 + Unripe
        price_rm = 15

    # Weight-based adjustments
    if weight > 550:
        price_rm += 5  # Premium for large fruit
    if weight < 350:
        price_rm -= 3  # Discount for small fruit

    # Convert to INR (1 RM ≈ 18.5 INR)
    price_inr = price_rm * 18.5

    return price_rm, price_inr


# ==========================================================================================================
# FUNCTION 9: MAIN PREDICTION SYSTEM (AGENTIC AI)
# ==========================================================================================================
# Purpose: Real-time mango quality assessment with comprehensive validation
# Validation Pipeline (7 checks):
#   1. Weight validation (300-700g)
#   2. Image loading verification
#   3. Blur detection (Laplacian variance)
#   4. Brightness validation (NEW - HSV V-channel)
#   5. Duplicate detection (SSIM)
#   6. Same mango verification (Histogram + SSIM + ORB)
#   7. Feature extraction & prediction
# Output: Ripeness, Grade, Price (RM + INR)
# ==========================================================================================================

def predict_mango(front, back, weight=None):
    """
    Comprehensive mango quality assessment system

    Validation Pipeline:
        1. Weight check
        2. Image loading
        3. Blur detection
        4. Brightness validation
        5. Duplicate detection
        6. Same mango verification
        7. Prediction

    Args:
        front: Path to front view image
        back: Path to back view image
        weight: Weight in grams (optional, will prompt if None)

    Returns:
        None (prints results to console)
    """

    print("\n" + "="*90)
    print("🍋 MANGO QUALITY ASSESSMENT SYSTEM".center(90))
    print("="*90)

    # -------------------------
    # VALIDATION 1: Weight
    # -------------------------
    if weight is None:
        try:
            weight = float(input("\n📝 Enter mango weight (grams): "))
        except ValueError:
            print("❌ VALIDATION FAILED: Invalid weight format")
            return

    print(f"\n[1/7] Weight Validation ({weight}g)...")
    if not check_weight(weight):
        print(f"    ❌ FAILED: Weight out of range (valid: 300-700g)")
        return
    print(f"    ✓ PASSED")

    # -------------------------
    # VALIDATION 2: Image Loading
    # -------------------------
    print(f"\n[2/7] Loading Images...")
    img1 = cv2.imread(front)
    img2 = cv2.imread(back)

    if img1 is None:
        print(f"    ❌ FAILED: Cannot load front image: {front}")
        return
    if img2 is None:
        print(f"    ❌ FAILED: Cannot load back image: {back}")
        return

    # Resize for processing
    img1 = cv2.resize(img1, (224, 224))
    img2 = cv2.resize(img2, (224, 224))
    print(f"    ✓ PASSED (Front: {front}, Back: {back})")

    # -------------------------
    # VALIDATION 3: Blur Detection
    # -------------------------
    print(f"\n[3/7] Blur Detection...")
    is_blur1, score1 = check_blur(img1)
    is_blur2, score2 = check_blur(img2)

    if is_blur1:
        print(f"    ❌ FAILED: Front image is blurry (score: {score1:.2f}, threshold: 35)")
        print(f"    💡 TIP: Ensure proper focus before capturing")
        return

    if is_blur2:
        print(f"    ❌ FAILED: Back image is blurry (score: {score2:.2f}, threshold: 35)")
        print(f"    💡 TIP: Ensure proper focus before capturing")
        return

    print(f"    ✓ PASSED (Front: {score1:.2f}, Back: {score2:.2f})")

    # -------------------------
    # VALIDATION 4: Brightness Check (NEW)
    # -------------------------
    print(f"\n[4/7] Brightness Validation...")
    b1_status, b1_val = check_brightness(img1)
    b2_status, b2_val = check_brightness(img2)

    if b1_status == "too_dark":
        print(f"    ❌ FAILED: Front image too dark (brightness: {b1_val:.1f}/255)")
        print(f"    💡 TIP: Increase lighting or use flash")
        return
    elif b1_status == "too_bright":
        print(f"    ❌ FAILED: Front image too bright (brightness: {b1_val:.1f}/255)")
        print(f"    💡 TIP: Reduce lighting or avoid direct sunlight")
        return

    if b2_status == "too_dark":
        print(f"    ❌ FAILED: Back image too dark (brightness: {b2_val:.1f}/255)")
        print(f"    💡 TIP: Increase lighting or use flash")
        return
    elif b2_status == "too_bright":
        print(f"    ❌ FAILED: Back image too bright (brightness: {b2_val:.1f}/255)")
        print(f"    💡 TIP: Reduce lighting or avoid direct sunlight")
        return

    print(f"    ✓ PASSED (Front: {b1_val:.1f}, Back: {b2_val:.1f})")

    # -------------------------
    # VALIDATION 5: Duplicate Detection
    # -------------------------
    print(f"\n[5/7] Duplicate Detection...")
    is_dup, dup_score = check_exact_duplicate(img1, img2)

    if is_dup:
        print(f"    ❌ FAILED: Images are duplicates (SSIM: {dup_score:.3f})")
        print(f"    💡 TIP: Upload DIFFERENT views (front AND back)")
        return

    print(f"    ✓ PASSED (SSIM: {dup_score:.3f})")

    # -------------------------
    # VALIDATION 6: Same Mango Verification
    # -------------------------
    print(f"\n[6/7] Same Mango Verification (Multi-Check)...")
    is_same, details = check_same_mango(img1, img2)

    if not is_same:
        print(f"    ❌ FAILED: Images appear to be from DIFFERENT mangoes")
        print(f"    📊 Check Results:")
        print(f"       - Histogram Correlation: {details['hist_sim']:.3f} (threshold: 0.85)")
        print(f"       - SSIM: {details['ssim']:.3f} (threshold: 0.75)")
        print(f"       - ORB Matches: {details['orb_matches']} (threshold: 8)")
        print(f"    💡 TIP: Ensure front and back are from THE SAME mango")
        return

    print(f"    ✓ PASSED")
    print(f"       - Histogram Correlation: {details['hist_sim']:.3f} ✓")
    print(f"       - SSIM: {details['ssim']:.3f} ✓")
    print(f"       - ORB Matches: {details['orb_matches']} ✓")

    # -------------------------
    # VALIDATION 7: Prediction
    # -------------------------
    print(f"\n[7/7] Running ML Prediction...")

    # Extract features
    f1 = extract_image_features(img1)
    f2 = extract_image_features(img2)
    feat = (f1 + f2) / 2

    # Prepare input
    X_input = np.hstack([feat, [weight]]).reshape(1, -1)
    X_input = final_scaler.transform(X_input)

    # Predict
    rip_pred = final_gb_rip.predict(X_input)
    grd_pred = final_gb_grd.predict(X_input)

    # Get probabilities for confidence
    rip_proba = final_gb_rip.predict_proba(X_input)[0]
    grd_proba = final_gb_grd.predict_proba(X_input)[0]

    # Decode predictions
    ripeness = rip_enc.inverse_transform(rip_pred)[0]
    grade = grd_enc.inverse_transform(grd_pred)[0]

    # Calculate confidence
    rip_confidence = max(rip_proba) * 100
    grd_confidence = max(grd_proba) * 100

    # Calculate prices (RM + INR)
    price_rm, price_inr = price_decision(grade, ripeness, weight)

    # -------------------------
    # Display Results
    # -------------------------
    print("\n" + "="*90)
    print("✅ ALL VALIDATIONS PASSED - ANALYSIS COMPLETE".center(90))
    print("="*90)

    print(f"\n{'QUALITY ASSESSMENT RESULTS':^90}")
    print("-"*90)
    print(f"{'Parameter':<30} {'Result':<30} {'Confidence':<30}")
    print("-"*90)
    print(f"{'🍋 Ripeness Stage':<30} {ripeness:<30} {f'{rip_confidence:.1f}%':<30}")
    print(f"{'🏆 Quality Grade':<30} {grade:<30} {f'{grd_confidence:.1f}%':<30}")
    print(f"{'⚖️  Weight':<30} {f'{weight}g':<30} {'-':<30}")
    print("-"*90)

    print(f"\n{'ESTIMATED MARKET PRICE':^90}")
    print("-"*90)
    print(f"{'Currency':<30} {'Price':<60}")
    print("-"*90)
    print(f"{'💵 Malaysian Ringgit (RM)':<30} {f'RM {price_rm:.2f}':<60}")
    print(f"{'💰 Indian Rupees (INR)':<30} {f'₹ {price_inr:.2f}':<60}")
    print("="*90)


# ==========================================================================================================
# STEP 6: SYSTEM READY
# ==========================================================================================================

print("\n" + "="*90)
print("[STEP 6/6] SYSTEM DEPLOYMENT READY".center(90))
print("="*90)

print("\n📋 GENERATED FILES:")
print("   ✓ stage1_front_view_features.xlsx")
print("   ✓ stage2_back_view_features.xlsx")
print("   ✓ stage3_combined_features.xlsx")
print("   ✓ stage4_fused_with_weight.xlsx")

print("\n🚀 USAGE:")
print("   predict_mango('/path/to/front.jpg', '/path/to/back.jpg', 450)")
print("\n   OR (interactive mode):")
print("   predict_mango('/path/to/front.jpg', '/path/to/back.jpg')")

print("\n" + "="*90)
print("🎯 SYSTEM READY FOR REAL-TIME MANGO QUALITY ASSESSMENT".center(90))
print("="*90)
```

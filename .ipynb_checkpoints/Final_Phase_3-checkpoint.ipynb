# Logistic Regression

```python
# ==========================================================================================================
# LOGISTIC REGRESSION - MULTIMODAL MULTI-VIEW MULTITASK MANGO GRADING SYSTEM
# ==========================================================================================================
#
# SYSTEM OVERVIEW:
# ┌─────────────────────────────────────────────────────────────────────────────────────────────────────┐
# │ Model: Logistic Regression (L2 Regularization)                                                     │
# │ Tasks: Ripeness (Binary) + Grade (Multiclass)                                                      │
# │ Fusion: Early, Late, Mid-Level, Hybrid                                                             │
# │ Multi-View: Front + Back images averaged                                                           │
# │ Features: 37 image features + 1 weight = 38 total                                                  │
# └─────────────────────────────────────────────────────────────────────────────────────────────────────┘
#
# Author: AI Engineer
# Date: 2026-03-22
# Version: 1.0
#
# ==========================================================================================================

import os
import cv2
import numpy as np
import pandas as pd
import warnings

from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

warnings.filterwarnings('ignore')

print("="*100)
print("LOGISTIC REGRESSION - MULTIMODAL MULTITASK FUSION PIPELINE".center(100))
print("="*100)


# ==========================================================================================================
# CONFIGURATION
# ==========================================================================================================

class Config:
    """Global configuration for the pipeline"""

    # File paths
    IMAGE_DIR = "images"
    EXCEL_PATH = "Harumanis_mango_weight_grade.xlsx"
    OUTPUT_FILE = "logistic_regression_fusion_results.xlsx"

    # Data split
    TEST_SIZE = 0.30
    RANDOM_STATE = 42

    # Logistic Regression parameters
    LR_C = 0.1                          # Regularization strength (smaller = stronger)
    LR_PENALTY = 'l2'                   # L2 regularization
    LR_MAX_ITER = 2000                  # Maximum iterations
    LR_SOLVER = 'lbfgs'                 # Solver (supports multinomial)
    LR_CLASS_WEIGHT = 'balanced'        # Handle class imbalance

    # Task-specific fusion weights
    RIPENESS_IMAGE_WEIGHT = 1.0         # Image importance for ripeness
    RIPENESS_WEIGHT_WEIGHT = 0.2        # Weight importance for ripeness

    GRADE_IMAGE_WEIGHT = 0.7            # Image importance for grade
    GRADE_WEIGHT_WEIGHT = 1.0           # Weight importance for grade


# ==========================================================================================================
# SECTION 1: FEATURE EXTRACTION
# ==========================================================================================================

def extract_image_features(img: np.ndarray) -> np.ndarray:
    """
    Extract 37 image features from mango image

    Feature Groups:
      - RGB statistics (6): mean and std for B, G, R channels
      - HSV statistics (6): mean and std for H, S, V channels
      - Hue histogram (16): color distribution for ripeness detection
      - GLCM texture (5): contrast, homogeneity, energy, correlation, dissimilarity
      - Color ratios (4): yellow ratio, green ratio, saturation stats, value stats

    Args:
        img: Input BGR image (OpenCV format)

    Returns:
        features: 1D numpy array of 37 features
    """

    # Resize to standard size
    img = cv2.resize(img, (224, 224))

    # ═══════════════════════════════════════════════════════════════════════════════
    # RGB Color Statistics (6 features)
    # ═══════════════════════════════════════════════════════════════════════════════
    rgb_mean = img.mean(axis=(0, 1))    # Mean of B, G, R
    rgb_std = img.std(axis=(0, 1))      # Std of B, G, R

    # ═══════════════════════════════════════════════════════════════════════════════
    # HSV Color Statistics (6 features)
    # ═══════════════════════════════════════════════════════════════════════════════
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0, 1))    # Mean of H, S, V
    hsv_std = hsv.std(axis=(0, 1))      # Std of H, S, V

    # ═══════════════════════════════════════════════════════════════════════════════
    # Hue Histogram (16 features)
    # ═══════════════════════════════════════════════════════════════════════════════
    # Captures color distribution - critical for ripeness detection
    hue = hsv[:, :, 0]
    hue_hist, _ = np.histogram(hue, bins=16, range=(0, 180))
    hue_hist = hue_hist / (hue_hist.sum() + 1e-10)  # Normalize

    # ═══════════════════════════════════════════════════════════════════════════════
    # GLCM Texture Features (5 features)
    # ═══════════════════════════════════════════════════════════════════════════════
    # Gray-Level Co-occurrence Matrix for texture analysis
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_quantized = (gray / 4).astype(np.uint8)  # Reduce to 64 levels

    glcm = graycomatrix(
        gray_quantized,
        distances=[1],
        angles=[0, np.pi/4, np.pi/2, 3*np.pi/4],
        levels=64,
        symmetric=True,
        normed=True
    )

    texture = [
        graycoprops(glcm, 'contrast').mean(),
        graycoprops(glcm, 'homogeneity').mean(),
        graycoprops(glcm, 'energy').mean(),
        graycoprops(glcm, 'correlation').mean(),
        graycoprops(glcm, 'dissimilarity').mean()
    ]

    # ═══════════════════════════════════════════════════════════════════════════════
    # Color Ratios (4 features) - Already included in HSV stats above
    # Total: 6 + 6 + 16 + 5 = 33 features
    # We'll add 4 more derived features to reach 37
    # ═══════════════════════════════════════════════════════════════════════════════

    # Yellow and green ratios (important for ripeness)
    yellow_mask = (hue >= 20) & (hue <= 35)
    green_mask = (hue >= 35) & (hue <= 85)

    yellow_ratio = yellow_mask.sum() / hue.size
    green_ratio = green_mask.sum() / hue.size

    # Saturation and value statistics
    sat_mean = hsv[:, :, 1].mean()
    sat_std = hsv[:, :, 1].std()
    val_mean = hsv[:, :, 2].mean()
    val_std = hsv[:, :, 2].std()

    # Combine all features
    all_features = np.hstack([
        rgb_mean,           # 3
        rgb_std,            # 3
        hsv_mean,           # 3
        hsv_std,            # 3
        hue_hist,           # 16
        texture,            # 5
        yellow_ratio,       # 1
        green_ratio,        # 1
        sat_mean,           # 1
        sat_std,            # 1
        val_mean,           # 1
        val_std             # 1
    ])

    return all_features  # Total: 37 features


# ==========================================================================================================
# SECTION 2: DATA LOADING & PREPROCESSING
# ==========================================================================================================

print("\n" + "="*100)
print("[STEP 1/6] LOADING DATASET & EXTRACTING FEATURES".center(100))
print("="*100)

# Load dataset
df = pd.read_excel(Config.EXCEL_PATH)
df["Fruit Grade"] = df["Fruit Grade"].astype(str)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

print(f"\n✓ Loaded {len(df)} records from Excel")

# Storage for features and labels
X_front = []        # Front view features
X_back = []         # Back view features
X_mean = []         # Averaged features
X_weight = []       # Weight features
y_ripeness = []     # Ripeness labels
y_grade = []        # Grade labels
groups = []         # Fruit IDs for grouping

loaded = 0
skipped = 0

# Process each fruit (requires 2 images: front + back)
for fruit_id, group in df.groupby("Fruit_ID"):

    if len(group) != 2:
        skipped += 1
        continue

    record1, record2 = group.iloc[0], group.iloc[1]

    # Load images
    img_path1 = os.path.join(Config.IMAGE_DIR, record1["Fruit No"])
    img_path2 = os.path.join(Config.IMAGE_DIR, record2["Fruit No"])

    img1 = cv2.imread(img_path1)
    img2 = cv2.imread(img_path2)

    if img1 is None or img2 is None:
        skipped += 1
        continue

    # Extract features
    features_front = extract_image_features(img1)
    features_back = extract_image_features(img2)
    features_averaged = (features_front + features_back) / 2

    # Get metadata
    weight_kg = record1["Mass(kg)"]
    ripeness = record1["Color_K-Yellow_P_Green"]
    grade = record1["Fruit Grade"]

    # Store
    X_front.append(features_front)
    X_back.append(features_back)
    X_mean.append(features_averaged)
    X_weight.append([weight_kg])
    y_ripeness.append(ripeness)
    y_grade.append(grade)
    groups.append(fruit_id)

    loaded += 1

# Convert to numpy arrays
X_front = np.array(X_front)
X_back = np.array(X_back)
X_mean = np.array(X_mean)
X_weight = np.array(X_weight)
groups = np.array(groups)

print(f"✓ Processed {loaded} mangoes (Skipped {skipped})")
print(f"✓ Feature shapes: Front={X_front.shape}, Back={X_back.shape}, Weight={X_weight.shape}")

# Encode labels
rip_enc = LabelEncoder()
grd_enc = LabelEncoder()

y_rip = rip_enc.fit_transform(y_ripeness)
y_grd = grd_enc.fit_transform(y_grade)

print(f"\n✓ Ripeness classes: {rip_enc.classes_}")
print(f"✓ Grade classes: {grd_enc.classes_}")


# ==========================================================================================================
# SECTION 3: DATA SPLITTING (GROUP-AWARE)
# ==========================================================================================================

print("\n" + "="*100)
print("[STEP 2/6] SPLITTING DATA (GROUP-AWARE)".center(100))
print("="*100)

# Group-aware split to prevent data leakage
# Same fruit ID will not appear in both train and test sets
gss = GroupShuffleSplit(n_splits=1, test_size=Config.TEST_SIZE, random_state=Config.RANDOM_STATE)
train_idx, test_idx = next(gss.split(X_mean, y_rip, groups))

# Split all feature sets
X_front_train, X_front_test = X_front[train_idx], X_front[test_idx]
X_back_train, X_back_test = X_back[train_idx], X_back[test_idx]
X_mean_train, X_mean_test = X_mean[train_idx], X_mean[test_idx]
X_weight_train, X_weight_test = X_weight[train_idx], X_weight[test_idx]

# Split labels
yr_train, yr_test = y_rip[train_idx], y_rip[test_idx]
yg_train, yg_test = y_grd[train_idx], y_grd[test_idx]

print(f"\n✓ Train samples: {len(train_idx)}")
print(f"✓ Test samples: {len(test_idx)}")
print(f"✓ Train/Test split: {(1-Config.TEST_SIZE)*100:.0f}/{Config.TEST_SIZE*100:.0f}")


# ==========================================================================================================
# SECTION 4: HELPER FUNCTIONS
# ==========================================================================================================

def train_logistic_regression(X_train, y_train, X_test, y_test, task='binary'):
    """
    Train and evaluate Logistic Regression model

    Args:
        X_train, y_train: Training data
        X_test, y_test: Test data
        task: 'binary' for ripeness, 'multiclass' for grade

    Returns:
        result: Dictionary with metrics
    """

    # Scale features
    scaler = RobustScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Configure Logistic Regression
    if task == 'multiclass':
        lr = LogisticRegression(
            C=Config.LR_C,
            penalty=Config.LR_PENALTY,
            max_iter=Config.LR_MAX_ITER,
            solver=Config.LR_SOLVER,
            class_weight=Config.LR_CLASS_WEIGHT,
            multi_class='multinomial',  # For multiclass
            random_state=Config.RANDOM_STATE
        )
    else:  # binary
        lr = LogisticRegression(
            C=Config.LR_C,
            penalty=Config.LR_PENALTY,
            max_iter=Config.LR_MAX_ITER,
            solver=Config.LR_SOLVER,
            class_weight=Config.LR_CLASS_WEIGHT,
            random_state=Config.RANDOM_STATE
        )

    # Train
    lr.fit(X_train_scaled, y_train)

    # Predict
    y_pred = lr.predict(X_test_scaled)

    # Calculate metrics
    result = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
        'recall': recall_score(y_test, y_pred, average='weighted', zero_division=0),
        'f1': f1_score(y_test, y_pred, average='weighted', zero_division=0),
        'model': lr,
        'scaler': scaler
    }

    return result


def get_probabilities(model, scaler, X):
    """Get probability predictions"""
    X_scaled = scaler.transform(X)
    return model.predict_proba(X_scaled)


# Storage for all results
all_results = []


# ==========================================================================================================
# FUSION 1: EARLY FUSION
# ==========================================================================================================
# Strategy: Combine image features + weight BEFORE training
# Task-specific weighting applied to emphasize different modalities
# ==========================================================================================================

print("\n" + "="*100)
print("[STEP 3/6] FUSION 1: EARLY FUSION".center(100))
print("="*100)

print("\n📊 Early Fusion Strategy:")
print("   - Concatenate: [Averaged Image Features (37)] + [Weight (1)] = 38 features")
print("   - Apply task-specific weighting:")
print(f"     • Ripeness: Image×{Config.RIPENESS_IMAGE_WEIGHT}, Weight×{Config.RIPENESS_WEIGHT_WEIGHT}")
print(f"     • Grade: Image×{Config.GRADE_IMAGE_WEIGHT}, Weight×{Config.GRADE_WEIGHT_WEIGHT}")

# ─────────────────────────────────────────────────────────────────────────────────
# Ripeness (Image-dominant)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[1/2] Training Ripeness Model (Early Fusion)...")

# Apply task-specific weights
X_early_rip_train = np.hstack([
    X_mean_train * Config.RIPENESS_IMAGE_WEIGHT,
    X_weight_train * Config.RIPENESS_WEIGHT_WEIGHT
])
X_early_rip_test = np.hstack([
    X_mean_test * Config.RIPENESS_IMAGE_WEIGHT,
    X_weight_test * Config.RIPENESS_WEIGHT_WEIGHT
])

# Train and evaluate
result_early_rip = train_logistic_regression(
    X_early_rip_train, yr_train,
    X_early_rip_test, yr_test,
    task='binary'
)

print(f"   ✓ Accuracy: {result_early_rip['accuracy']:.4f}")
print(f"   ✓ F1-Score: {result_early_rip['f1']:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Early Fusion',
    'Task': 'Ripeness',
    'Accuracy': round(result_early_rip['accuracy'], 6),
    'Precision': round(result_early_rip['precision'], 6),
    'Recall': round(result_early_rip['recall'], 6),
    'F1': round(result_early_rip['f1'], 6)
})

# ─────────────────────────────────────────────────────────────────────────────────
# Grade (Weight-dominant for physical quality)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[2/2] Training Grade Model (Early Fusion)...")

# Apply task-specific weights
X_early_grd_train = np.hstack([
    X_mean_train * Config.GRADE_IMAGE_WEIGHT,
    X_weight_train * Config.GRADE_WEIGHT_WEIGHT
])
X_early_grd_test = np.hstack([
    X_mean_test * Config.GRADE_IMAGE_WEIGHT,
    X_weight_test * Config.GRADE_WEIGHT_WEIGHT
])

# Train and evaluate
result_early_grd = train_logistic_regression(
    X_early_grd_train, yg_train,
    X_early_grd_test, yg_test,
    task='multiclass'
)

print(f"   ✓ Accuracy: {result_early_grd['accuracy']:.4f}")
print(f"   ✓ F1-Score: {result_early_grd['f1']:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Early Fusion',
    'Task': 'Grade',
    'Accuracy': round(result_early_grd['accuracy'], 6),
    'Precision': round(result_early_grd['precision'], 6),
    'Recall': round(result_early_grd['recall'], 6),
    'F1': round(result_early_grd['f1'], 6)
})

print("\n✅ Early Fusion Complete")


# ==========================================================================================================
# FUSION 2: LATE FUSION
# ==========================================================================================================
# Strategy: Train separate models on each view, then combine predictions
# Models: Front-only, Back-only, Weight-only
# Combination: Average probabilities, then argmax
# ==========================================================================================================

print("\n" + "="*100)
print("[STEP 4/6] FUSION 2: LATE FUSION".center(100))
print("="*100)

print("\n📊 Late Fusion Strategy:")
print("   - Train 3 separate Logistic Regression models:")
print("     1. Front image features only (37 features)")
print("     2. Back image features only (37 features)")
print("     3. Weight only (1 feature)")
print("   - Combine via probability averaging")

def late_fusion_predict(models_with_scalers, X_front, X_back, X_weight):
    """
    Combine predictions from multiple models via probability averaging

    Args:
        models_with_scalers: List of (model, scaler, X_data) tuples
        X_front, X_back, X_weight: Test data

    Returns:
        y_pred: Final predictions
    """

    all_probas = []

    for model, scaler, X_data in models_with_scalers:
        proba = get_probabilities(model, scaler, X_data)
        all_probas.append(proba)

    # Average probabilities
    avg_proba = np.mean(all_probas, axis=0)

    # Argmax to get final prediction
    y_pred = np.argmax(avg_proba, axis=1)

    return y_pred


# ─────────────────────────────────────────────────────────────────────────────────
# Ripeness (Late Fusion)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[1/2] Training Ripeness Models (Late Fusion)...")

# Train 3 separate models
print("   Training Front-only model...", end=' ')
result_front_rip = train_logistic_regression(X_front_train, yr_train, X_front_test, yr_test, 'binary')
print("✓")

print("   Training Back-only model...", end=' ')
result_back_rip = train_logistic_regression(X_back_train, yr_train, X_back_test, yr_test, 'binary')
print("✓")

print("   Training Weight-only model...", end=' ')
result_weight_rip = train_logistic_regression(X_weight_train, yr_train, X_weight_test, yr_test, 'binary')
print("✓")

# Combine predictions
models_rip = [
    (result_front_rip['model'], result_front_rip['scaler'], X_front_test),
    (result_back_rip['model'], result_back_rip['scaler'], X_back_test),
    (result_weight_rip['model'], result_weight_rip['scaler'], X_weight_test)
]

yr_pred_late = late_fusion_predict(models_rip, X_front_test, X_back_test, X_weight_test)

# Calculate metrics
late_rip_acc = accuracy_score(yr_test, yr_pred_late)
late_rip_prec = precision_score(yr_test, yr_pred_late, average='weighted', zero_division=0)
late_rip_rec = recall_score(yr_test, yr_pred_late, average='weighted', zero_division=0)
late_rip_f1 = f1_score(yr_test, yr_pred_late, average='weighted', zero_division=0)

print(f"   ✓ Combined Accuracy: {late_rip_acc:.4f}")
print(f"   ✓ Combined F1-Score: {late_rip_f1:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Late Fusion',
    'Task': 'Ripeness',
    'Accuracy': round(late_rip_acc, 6),
    'Precision': round(late_rip_prec, 6),
    'Recall': round(late_rip_rec, 6),
    'F1': round(late_rip_f1, 6)
})

# ─────────────────────────────────────────────────────────────────────────────────
# Grade (Late Fusion)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[2/2] Training Grade Models (Late Fusion)...")

# Train 3 separate models
print("   Training Front-only model...", end=' ')
result_front_grd = train_logistic_regression(X_front_train, yg_train, X_front_test, yg_test, 'multiclass')
print("✓")

print("   Training Back-only model...", end=' ')
result_back_grd = train_logistic_regression(X_back_train, yg_train, X_back_test, yg_test, 'multiclass')
print("✓")

print("   Training Weight-only model...", end=' ')
result_weight_grd = train_logistic_regression(X_weight_train, yg_train, X_weight_test, yg_test, 'multiclass')
print("✓")

# Combine predictions
models_grd = [
    (result_front_grd['model'], result_front_grd['scaler'], X_front_test),
    (result_back_grd['model'], result_back_grd['scaler'], X_back_test),
    (result_weight_grd['model'], result_weight_grd['scaler'], X_weight_test)
]

yg_pred_late = late_fusion_predict(models_grd, X_front_test, X_back_test, X_weight_test)

# Calculate metrics
late_grd_acc = accuracy_score(yg_test, yg_pred_late)
late_grd_prec = precision_score(yg_test, yg_pred_late, average='weighted', zero_division=0)
late_grd_rec = recall_score(yg_test, yg_pred_late, average='weighted', zero_division=0)
late_grd_f1 = f1_score(yg_test, yg_pred_late, average='weighted', zero_division=0)

print(f"   ✓ Combined Accuracy: {late_grd_acc:.4f}")
print(f"   ✓ Combined F1-Score: {late_grd_f1:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Late Fusion',
    'Task': 'Grade',
    'Accuracy': round(late_grd_acc, 6),
    'Precision': round(late_grd_prec, 6),
    'Recall': round(late_grd_rec, 6),
    'F1': round(late_grd_f1, 6)
})

print("\n✅ Late Fusion Complete")


# ==========================================================================================================
# FUSION 3: MID-LEVEL FUSION
# ==========================================================================================================
# Strategy: Two-stage learning
# Stage 1: Train on image features, extract decision values/probabilities
# Stage 2: Use extracted features + weight as input for final model
# ==========================================================================================================

print("\n" + "="*100)
print("[STEP 5/6] FUSION 3: MID-LEVEL FUSION".center(100))
print("="*100)

print("\n📊 Mid-Level Fusion Strategy:")
print("   - Stage 1: Train Logistic Regression on image features")
print("   - Extract: decision_function() values as intermediate features")
print("   - Stage 2: Concatenate with weight, train final model")

# ─────────────────────────────────────────────────────────────────────────────────
# Ripeness (Mid-Level Fusion)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[1/2] Training Ripeness Model (Mid-Level Fusion)...")

# Stage 1: Train on image features
print("   Stage 1: Training on image features...", end=' ')
scaler_stage1_rip = RobustScaler()
X_mean_train_s1 = scaler_stage1_rip.fit_transform(X_mean_train)
X_mean_test_s1 = scaler_stage1_rip.transform(X_mean_test)

lr_stage1_rip = LogisticRegression(
    C=Config.LR_C,
    penalty=Config.LR_PENALTY,
    max_iter=Config.LR_MAX_ITER,
    solver=Config.LR_SOLVER,
    class_weight=Config.LR_CLASS_WEIGHT,
    random_state=Config.RANDOM_STATE
)
lr_stage1_rip.fit(X_mean_train_s1, yr_train)
print("✓")

# Extract decision function values (confidence scores)
print("   Extracting decision features...", end=' ')
decision_train_rip = lr_stage1_rip.decision_function(X_mean_train_s1)
decision_test_rip = lr_stage1_rip.decision_function(X_mean_test_s1)

# Reshape if 1D (binary classification returns 1D array)
if decision_train_rip.ndim == 1:
    decision_train_rip = decision_train_rip.reshape(-1, 1)
    decision_test_rip = decision_test_rip.reshape(-1, 1)
print("✓")

# Stage 2: Combine with weight
print("   Stage 2: Training final model...", end=' ')
X_mid_rip_train = np.hstack([decision_train_rip, X_weight_train])
X_mid_rip_test = np.hstack([decision_test_rip, X_weight_test])

result_mid_rip = train_logistic_regression(
    X_mid_rip_train, yr_train,
    X_mid_rip_test, yr_test,
    task='binary'
)
print("✓")

print(f"   ✓ Accuracy: {result_mid_rip['accuracy']:.4f}")
print(f"   ✓ F1-Score: {result_mid_rip['f1']:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Mid-Level Fusion',
    'Task': 'Ripeness',
    'Accuracy': round(result_mid_rip['accuracy'], 6),
    'Precision': round(result_mid_rip['precision'], 6),
    'Recall': round(result_mid_rip['recall'], 6),
    'F1': round(result_mid_rip['f1'], 6)
})

# ─────────────────────────────────────────────────────────────────────────────────
# Grade (Mid-Level Fusion)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[2/2] Training Grade Model (Mid-Level Fusion)...")

# Stage 1: Train on image features
print("   Stage 1: Training on image features...", end=' ')
scaler_stage1_grd = RobustScaler()
X_mean_train_s1_g = scaler_stage1_grd.fit_transform(X_mean_train)
X_mean_test_s1_g = scaler_stage1_grd.transform(X_mean_test)

lr_stage1_grd = LogisticRegression(
    C=Config.LR_C,
    penalty=Config.LR_PENALTY,
    max_iter=Config.LR_MAX_ITER,
    solver=Config.LR_SOLVER,
    class_weight=Config.LR_CLASS_WEIGHT,
    multi_class='multinomial',
    random_state=Config.RANDOM_STATE
)
lr_stage1_grd.fit(X_mean_train_s1_g, yg_train)
print("✓")

# Extract decision function values
print("   Extracting decision features...", end=' ')
decision_train_grd = lr_stage1_grd.decision_function(X_mean_train_s1_g)
decision_test_grd = lr_stage1_grd.decision_function(X_mean_test_s1_g)
print("✓")

# Stage 2: Combine with weight
print("   Stage 2: Training final model...", end=' ')
X_mid_grd_train = np.hstack([decision_train_grd, X_weight_train])
X_mid_grd_test = np.hstack([decision_test_grd, X_weight_test])

result_mid_grd = train_logistic_regression(
    X_mid_grd_train, yg_train,
    X_mid_grd_test, yg_test,
    task='multiclass'
)
print("✓")

print(f"   ✓ Accuracy: {result_mid_grd['accuracy']:.4f}")
print(f"   ✓ F1-Score: {result_mid_grd['f1']:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Mid-Level Fusion',
    'Task': 'Grade',
    'Accuracy': round(result_mid_grd['accuracy'], 6),
    'Precision': round(result_mid_grd['precision'], 6),
    'Recall': round(result_mid_grd['recall'], 6),
    'F1': round(result_mid_grd['f1'], 6)
})

print("\n✅ Mid-Level Fusion Complete")


# ==========================================================================================================
# FUSION 4: HYBRID FUSION
# ==========================================================================================================
# Strategy: Ensemble of Early + Late + Mid-Level fusion predictions
# Combination: Average probabilities from all three approaches
# ==========================================================================================================

print("\n" + "="*100)
print("[STEP 6/6] FUSION 4: HYBRID FUSION".center(100))
print("="*100)

print("\n📊 Hybrid Fusion Strategy:")
print("   - Combine predictions from:")
print("     1. Early Fusion")
print("     2. Late Fusion (already combined)")
print("     3. Mid-Level Fusion")
print("   - Method: Average probabilities → Argmax")

# ─────────────────────────────────────────────────────────────────────────────────
# Ripeness (Hybrid Fusion)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[1/2] Combining Ripeness Predictions (Hybrid Fusion)...")

# Get probabilities from each fusion method
print("   Extracting probabilities from each fusion...", end=' ')

# Early fusion probabilities
proba_early_rip = get_probabilities(
    result_early_rip['model'],
    result_early_rip['scaler'],
    X_early_rip_test
)

# Late fusion probabilities (re-compute from individual models)
proba_late_rip = np.mean([
    get_probabilities(m, s, X)
    for m, s, X in models_rip
], axis=0)

# Mid-level fusion probabilities
proba_mid_rip = get_probabilities(
    result_mid_rip['model'],
    result_mid_rip['scaler'],
    X_mid_rip_test
)

print("✓")

# Average all probabilities
print("   Averaging probabilities...", end=' ')
proba_hybrid_rip = (proba_early_rip + proba_late_rip + proba_mid_rip) / 3
yr_pred_hybrid = np.argmax(proba_hybrid_rip, axis=1)
print("✓")

# Calculate metrics
hybrid_rip_acc = accuracy_score(yr_test, yr_pred_hybrid)
hybrid_rip_prec = precision_score(yr_test, yr_pred_hybrid, average='weighted', zero_division=0)
hybrid_rip_rec = recall_score(yr_test, yr_pred_hybrid, average='weighted', zero_division=0)
hybrid_rip_f1 = f1_score(yr_test, yr_pred_hybrid, average='weighted', zero_division=0)

print(f"   ✓ Accuracy: {hybrid_rip_acc:.4f}")
print(f"   ✓ F1-Score: {hybrid_rip_f1:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Hybrid Fusion',
    'Task': 'Ripeness',
    'Accuracy': round(hybrid_rip_acc, 6),
    'Precision': round(hybrid_rip_prec, 6),
    'Recall': round(hybrid_rip_rec, 6),
    'F1': round(hybrid_rip_f1, 6)
})

# ─────────────────────────────────────────────────────────────────────────────────
# Grade (Hybrid Fusion)
# ─────────────────────────────────────────────────────────────────────────────────

print("\n[2/2] Combining Grade Predictions (Hybrid Fusion)...")

# Get probabilities from each fusion method
print("   Extracting probabilities from each fusion...", end=' ')

# Early fusion probabilities
proba_early_grd = get_probabilities(
    result_early_grd['model'],
    result_early_grd['scaler'],
    X_early_grd_test
)

# Late fusion probabilities
proba_late_grd = np.mean([
    get_probabilities(m, s, X)
    for m, s, X in models_grd
], axis=0)

# Mid-level fusion probabilities
proba_mid_grd = get_probabilities(
    result_mid_grd['model'],
    result_mid_grd['scaler'],
    X_mid_grd_test
)

print("✓")

# Average all probabilities
print("   Averaging probabilities...", end=' ')
proba_hybrid_grd = (proba_early_grd + proba_late_grd + proba_mid_grd) / 3
yg_pred_hybrid = np.argmax(proba_hybrid_grd, axis=1)
print("✓")

# Calculate metrics
hybrid_grd_acc = accuracy_score(yg_test, yg_pred_hybrid)
hybrid_grd_prec = precision_score(yg_test, yg_pred_hybrid, average='weighted', zero_division=0)
hybrid_grd_rec = recall_score(yg_test, yg_pred_hybrid, average='weighted', zero_division=0)
hybrid_grd_f1 = f1_score(yg_test, yg_pred_hybrid, average='weighted', zero_division=0)

print(f"   ✓ Accuracy: {hybrid_grd_acc:.4f}")
print(f"   ✓ F1-Score: {hybrid_grd_f1:.4f}")

# Store results
all_results.append({
    'Model': 'Logistic Regression',
    'Fusion': 'Hybrid Fusion',
    'Task': 'Grade',
    'Accuracy': round(hybrid_grd_acc, 6),
    'Precision': round(hybrid_grd_prec, 6),
    'Recall': round(hybrid_grd_rec, 6),
    'F1': round(hybrid_grd_f1, 6)
})

print("\n✅ Hybrid Fusion Complete")


# ==========================================================================================================
# SECTION 5: RESULTS SUMMARY & EXPORT
# ==========================================================================================================

print("\n" + "="*100)
print("FINAL RESULTS SUMMARY".center(100))
print("="*100)

# Convert results to DataFrame
results_df = pd.DataFrame(all_results)

# Display results
print("\n" + results_df.to_string(index=False))

# Save to Excel
results_df.to_excel(Config.OUTPUT_FILE, index=False)

print("\n" + "="*100)
print(f"✅ RESULTS SAVED TO: {Config.OUTPUT_FILE}".center(100))
print("="*100)

# Display summary statistics
print("\n📊 SUMMARY STATISTICS:")
print("-"*100)

for task in ['Ripeness', 'Grade']:
    task_results = results_df[results_df['Task'] == task]

    print(f"\n{task.upper()}:")
    print(f"   Best F1-Score: {task_results['F1'].max():.4f} ({task_results.loc[task_results['F1'].idxmax(), 'Fusion']})")
    print(f"   Avg F1-Score:  {task_results['F1'].mean():.4f}")
    print(f"   Best Accuracy: {task_results['Accuracy'].max():.4f} ({task_results.loc[task_results['Accuracy'].idxmax(), 'Fusion']})")

print("\n" + "="*100)
print("✅ LOGISTIC REGRESSION PIPELINE COMPLETE".center(100))
print("="*100)
```

# SVM (RBF)

```python
import os
import cv2
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit, cross_val_score
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix
)

print("="*70)
print("SVM (RBF) - MULTIMODAL FUSION PIPELINE (FIXED)")
print("="*70)

# ======================================================
# PATHS
# ======================================================
IMAGE_DIR  = "images"
EXCEL_PATH = "Harumanis_mango_weight_grade.xlsx"

# ======================================================
# LOAD DATA
# ======================================================
print("\n[Step 1/10] Loading Dataset...")
df = pd.read_excel(EXCEL_PATH)
df["Fruit Grade"] = df["Fruit Grade"].astype(str)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

print(f"✓ Loaded {len(df)} records")

# ======================================================
# FEATURE EXTRACTION
# ======================================================
print("\n[Step 2/10] Defining Feature Extraction...")

def extract_image_features(img):
    """Extract exactly 20 features"""
    img = cv2.resize(img, (224, 224))
    rgb_mean = img.mean(axis=(0,1))
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0,1))
    hsv_std  = hsv.std(axis=(0,1))
    hue = hsv[:, :, 0]
    yellow_mask = (hue >= 20) & (hue <= 35)
    green_mask  = (hue >= 35) & (hue <= 85)
    yellow_ratio = yellow_mask.sum() / hue.size
    green_ratio  = green_mask.sum() / hue.size
    sat_mean = hsv[:,:,1].mean()
    sat_std  = hsv[:,:,1].std()
    val_mean = hsv[:,:,2].mean()
    val_std  = hsv[:,:,2].std()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = (gray / 4).astype(np.uint8)
    glcm = graycomatrix(gray, distances=[1], angles=[0, np.pi/4, np.pi/2, 3*np.pi/4],
                        levels=64, symmetric=True, normed=True)
    contrast = graycoprops(glcm, 'contrast').mean()
    homogeneity = graycoprops(glcm, 'homogeneity').mean()
    energy = graycoprops(glcm, 'energy').mean()
    correlation = graycoprops(glcm, 'correlation').mean()
    dissimilarity = graycoprops(glcm, 'dissimilarity').mean()
    return np.hstack([rgb_mean, hsv_mean, hsv_std, yellow_ratio, green_ratio,
                      sat_mean, sat_std, val_mean, val_std, contrast,
                      homogeneity, energy, correlation, dissimilarity])

FEATURE_NAMES = ['rgb_b', 'rgb_g', 'rgb_r', 'hsv_h', 'hsv_s', 'hsv_v',
                 'hsv_std_h', 'hsv_std_s', 'hsv_std_v', 'yellow_ratio',
                 'green_ratio', 'sat_mean', 'sat_std', 'val_mean', 'val_std',
                 'contrast', 'homogeneity', 'energy', 'correlation', 'dissimilarity']

print("✓ Feature extractor defined")

# ======================================================
# BUILD DATASET
# ======================================================
print("\n[Step 3/10] Loading Images...")

front_records, back_records, combined_records = [], [], []
y_ripeness, y_grade, groups = [], [], []
loaded, skipped = 0, 0

for fruit_id, g in df.groupby("Fruit_ID"):
    if len(g) != 2:
        skipped += 1
        continue
    r1, r2 = g.iloc[0], g.iloc[1]
    img1 = cv2.imread(os.path.join(IMAGE_DIR, r1["Fruit No"]))
    img2 = cv2.imread(os.path.join(IMAGE_DIR, r2["Fruit No"]))
    if img1 is None or img2 is None:
        skipped += 1
        continue
    f1 = extract_image_features(img1)
    f2 = extract_image_features(img2)
    weight = r1["Mass(kg)"]
    ripeness = r1["Color_K-Yellow_P_Green"]
    grade = r1["Fruit Grade"]

    front_row = {"Fruit_ID": fruit_id, "Image": r1["Fruit No"],
                 "Weight_kg": weight, "Ripeness": ripeness, "Grade": grade}
    for i, name in enumerate(FEATURE_NAMES):
        front_row[name] = f1[i]
    front_records.append(front_row)

    back_row = {"Fruit_ID": fruit_id, "Image": r2["Fruit No"],
                "Weight_kg": weight, "Ripeness": ripeness, "Grade": grade}
    for i, name in enumerate(FEATURE_NAMES):
        back_row[name] = f2[i]
    back_records.append(back_row)

    combined_row = {"Fruit_ID": fruit_id, "Front_Image": r1["Fruit No"],
                    "Back_Image": r2["Fruit No"], "Weight_kg": weight,
                    "Ripeness": ripeness, "Grade": grade}
    for i, name in enumerate(FEATURE_NAMES):
        combined_row[f"front_{name}"] = f1[i]
        combined_row[f"back_{name}"] = f2[i]
    combined_records.append(combined_row)

    y_ripeness.append(ripeness)
    y_grade.append(grade)
    groups.append(fruit_id)
    loaded += 1

print(f"✓ Loaded: {loaded} | Skipped: {skipped}")

# ======================================================
# SAVE STAGE EXCELS
# ======================================================
print("\n[Step 4/10] Saving Stage Files...")

pd.DataFrame(front_records).to_excel("svm_stage1_front.xlsx", index=False)
pd.DataFrame(back_records).to_excel("svm_stage2_back.xlsx", index=False)
pd.DataFrame(combined_records).to_excel("svm_stage3_combined.xlsx", index=False)

fused_records = []
for row in combined_records:
    fused_row = {"Fruit_ID": row["Fruit_ID"], "Front_Image": row["Front_Image"],
                 "Back_Image": row["Back_Image"], "Weight_kg": row["Weight_kg"],
                 "Ripeness": row["Ripeness"], "Grade": row["Grade"]}
    for name in FEATURE_NAMES:
        fused_row[f"front_{name}"] = row[f"front_{name}"]
        fused_row[f"back_{name}"] = row[f"back_{name}"]
        fused_row[f"fused_{name}"] = (row[f"front_{name}"] + row[f"back_{name}"]) / 2
    fused_row["physical_weight"] = row["Weight_kg"]
    fused_records.append(fused_row)
pd.DataFrame(fused_records).to_excel("svm_stage4_fused.xlsx", index=False)

print("✓ Stage files saved")

# ======================================================
# PREPARE ARRAYS
# ======================================================
print("\n[Step 5/10] Preparing Arrays...")

X_front = np.array([[row[f"front_{n}"] for n in FEATURE_NAMES] for row in combined_records])
X_back = np.array([[row[f"back_{n}"] for n in FEATURE_NAMES] for row in combined_records])
X_weight = np.array([[row["Weight_kg"]] for row in combined_records])
X_mean = (X_front + X_back) / 2
groups = np.array(groups)

rip_enc = LabelEncoder()
grd_enc = LabelEncoder()
yr = rip_enc.fit_transform(y_ripeness)
yg = grd_enc.fit_transform(y_grade)

print(f"✓ Ripeness classes: {rip_enc.classes_}")
print(f"✓ Grade classes: {grd_enc.classes_}")

# ======================================================
# SPLIT
# ======================================================
print("\n[Step 6/10] Splitting...")

gss = GroupShuffleSplit(n_splits=1, test_size=0.3, random_state=42)  # Increased test size
tr_idx, te_idx = next(gss.split(X_front, yr, groups))

X_front_tr, X_front_te = X_front[tr_idx], X_front[te_idx]
X_back_tr, X_back_te = X_back[tr_idx], X_back[te_idx]
X_weight_tr, X_weight_te = X_weight[tr_idx], X_weight[te_idx]
X_mean_tr, X_mean_te = X_mean[tr_idx], X_mean[te_idx]
yr_tr, yr_te = yr[tr_idx], yr[te_idx]
yg_tr, yg_te = yg[tr_idx], yg[te_idx]

print(f"✓ Train: {len(tr_idx)} | Test: {len(te_idx)}")

# ======================================================
# HELPER FUNCTION
# ======================================================
def train_eval_svm(X_tr, X_te, y_tr, y_te, C=0.1, gamma='scale'):
    scaler = RobustScaler()
    X_tr_s = scaler.fit_transform(X_tr)
    X_te_s = scaler.transform(X_te)
    svm = SVC(kernel='rbf', C=C, gamma=gamma, class_weight='balanced',
              random_state=42, probability=True)
    svm.fit(X_tr_s, y_tr)
    y_tr_pred = svm.predict(X_tr_s)
    y_te_pred = svm.predict(X_te_s)
    train_acc = accuracy_score(y_tr, y_tr_pred)
    test_acc = accuracy_score(y_te, y_te_pred)
    return {
        'train_acc': train_acc,
        'test_acc': test_acc,
        'gap': train_acc - test_acc,
        'accuracy': test_acc,
        'precision': precision_score(y_te, y_te_pred, average='weighted', zero_division=0),
        'recall': recall_score(y_te, y_te_pred, average='weighted', zero_division=0),
        'f1': f1_score(y_te, y_te_pred, average='weighted', zero_division=0)
    }

# ======================================================
# FUSION 1: EARLY FUSION
# ======================================================
print("\n" + "="*70)
print("FUSION 1: EARLY FUSION")
print("="*70)

X_early_rip_tr = np.hstack([X_mean_tr * 1.0, X_weight_tr * 0.2])
X_early_rip_te = np.hstack([X_mean_te * 1.0, X_weight_te * 0.2])
X_early_grd_tr = np.hstack([X_mean_tr * 0.7, X_weight_tr * 1.0])
X_early_grd_te = np.hstack([X_mean_te * 0.7, X_weight_te * 1.0])

print("\nTuning Ripeness...")
C_vals = [0.01, 0.05, 0.1, 0.5]  # Stronger regularization
gamma_vals = ['scale', 'auto']

best_rip_f1, best_rip_params, best_rip_result = 0, {}, {}
total = len(C_vals) * len(gamma_vals)
current = 0

for C in C_vals:
    for gamma in gamma_vals:
        current += 1
        print(f"  [{current}/{total}] C={C}, gamma={gamma}", end='')
        result = train_eval_svm(X_early_rip_tr, X_early_rip_te, yr_tr, yr_te, C, gamma)
        print(f" → F1={result['f1']:.4f}, Gap={result['gap']:.3f}")
        if result['f1'] > best_rip_f1:
            best_rip_f1 = result['f1']
            best_rip_params = {'C': C, 'gamma': gamma}
            best_rip_result = result

print(f"\n✓ Best: C={best_rip_params['C']}, gamma={best_rip_params['gamma']}, F1={best_rip_f1:.4f}")

print("\nTuning Grade...")
best_grd_f1, best_grd_params, best_grd_result = 0, {}, {}
current = 0

for C in C_vals:
    for gamma in gamma_vals:
        current += 1
        print(f"  [{current}/{total}] C={C}, gamma={gamma}", end='')
        result = train_eval_svm(X_early_grd_tr, X_early_grd_te, yg_tr, yg_te, C, gamma)
        print(f" → F1={result['f1']:.4f}, Gap={result['gap']:.3f}")
        if result['f1'] > best_grd_f1:
            best_grd_f1 = result['f1']
            best_grd_params = {'C': C, 'gamma': gamma}
            best_grd_result = result

print(f"\n✓ Best: C={best_grd_params['C']}, gamma={best_grd_params['gamma']}, F1={best_grd_f1:.4f}")

# FIXED COLUMN NAMES - NO TRUNCATION
early_results = pd.DataFrame([
    {
        'Fusion_Type': 'Early Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Ripeness',
        'Best_C': best_rip_params['C'],
        'Best_Gamma': best_rip_params['gamma'],
        'Train_Acc': round(best_rip_result['train_acc'], 6),
        'Test_Acc': round(best_rip_result['test_acc'], 6),
        'Overfit_Gap': round(best_rip_result['gap'], 6),
        'Accuracy': round(best_rip_result['accuracy'], 6),
        'Precision': round(best_rip_result['precision'], 6),
        'Recall': round(best_rip_result['recall'], 6),
        'F1_Score': round(best_rip_result['f1'], 6)
    },
    {
        'Fusion_Type': 'Early Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Grade',
        'Best_C': best_grd_params['C'],
        'Best_Gamma': best_grd_params['gamma'],
        'Train_Acc': round(best_grd_result['train_acc'], 6),
        'Test_Acc': round(best_grd_result['test_acc'], 6),
        'Overfit_Gap': round(best_grd_result['gap'], 6),
        'Accuracy': round(best_grd_result['accuracy'], 6),
        'Precision': round(best_grd_result['precision'], 6),
        'Recall': round(best_grd_result['recall'], 6),
        'F1_Score': round(best_grd_result['f1'], 6)
    }
])

early_results.to_excel("svm_early_fusion_results.xlsx", index=False)
print("\n✅ Early Fusion saved!")

# ======================================================
# FUSION 2: LATE FUSION
# ======================================================
print("\n" + "="*70)
print("FUSION 2: LATE FUSION")
print("="*70)

def late_fusion(X_front, X_back, X_weight, y_train, y_test):
    models_data = [('Front', X_front), ('Back', X_back), ('Weight', X_weight)]
    probas = []
    for name, X in models_data:
        scaler = RobustScaler()
        X_tr_s = scaler.fit_transform(X[0])
        X_te_s = scaler.transform(X[1])
        svm = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced',
                  random_state=42, probability=True)
        svm.fit(X_tr_s, y_train)
        probas.append(svm.predict_proba(X_te_s))
    avg_proba = np.mean(probas, axis=0)
    return np.argmax(avg_proba, axis=1)

print("Training Ripeness...")
yr_pred_late = late_fusion((X_front_tr, X_front_te), (X_back_tr, X_back_te),
                           (X_weight_tr, X_weight_te), yr_tr, yr_te)

print("Training Grade...")
yg_pred_late = late_fusion((X_front_tr, X_front_te), (X_back_tr, X_back_te),
                           (X_weight_tr, X_weight_te), yg_tr, yg_te)

late_results = pd.DataFrame([
    {
        'Fusion_Type': 'Late Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Ripeness',
        'Strategy': '3 Models Probability Avg',
        'Accuracy': round(accuracy_score(yr_te, yr_pred_late), 6),
        'Precision': round(precision_score(yr_te, yr_pred_late, average='weighted', zero_division=0), 6),
        'Recall': round(recall_score(yr_te, yr_pred_late, average='weighted', zero_division=0), 6),
        'F1_Score': round(f1_score(yr_te, yr_pred_late, average='weighted', zero_division=0), 6)
    },
    {
        'Fusion_Type': 'Late Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Grade',
        'Strategy': '3 Models Probability Avg',
        'Accuracy': round(accuracy_score(yg_te, yg_pred_late), 6),
        'Precision': round(precision_score(yg_te, yg_pred_late, average='weighted', zero_division=0), 6),
        'Recall': round(recall_score(yg_te, yg_pred_late, average='weighted', zero_division=0), 6),
        'F1_Score': round(f1_score(yg_te, yg_pred_late, average='weighted', zero_division=0), 6)
    }
])

late_results.to_excel("svm_late_fusion_results.xlsx", index=False)
print("\n✅ Late Fusion saved!")

# ======================================================
# FUSION 3: MID-LEVEL
# ======================================================
print("\n" + "="*70)
print("FUSION 3: MID-LEVEL FUSION")
print("="*70)

scaler_mid = RobustScaler()
X_mean_tr_s = scaler_mid.fit_transform(X_mean_tr)
X_mean_te_s = scaler_mid.transform(X_mean_te)

svm_img_rip = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced', random_state=42)
svm_img_rip.fit(X_mean_tr_s, yr_tr)
img_feat_tr_rip = svm_img_rip.decision_function(X_mean_tr_s).reshape(-1, 1)
img_feat_te_rip = svm_img_rip.decision_function(X_mean_te_s).reshape(-1, 1)

X_mid_rip_tr = np.hstack([img_feat_tr_rip, X_weight_tr])
X_mid_rip_te = np.hstack([img_feat_te_rip, X_weight_te])

scaler_mid_final = RobustScaler()
X_mid_rip_tr_s = scaler_mid_final.fit_transform(X_mid_rip_tr)
X_mid_rip_te_s = scaler_mid_final.transform(X_mid_rip_te)

svm_mid_rip = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced', random_state=42)
svm_mid_rip.fit(X_mid_rip_tr_s, yr_tr)
yr_pred_mid = svm_mid_rip.predict(X_mid_rip_te_s)
yr_train_pred_mid = svm_mid_rip.predict(X_mid_rip_tr_s)

print(f"✓ Ripeness - F1={f1_score(yr_te, yr_pred_mid, average='weighted', zero_division=0):.4f}")

# Grade
svm_img_grd = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced', random_state=42)
svm_img_grd.fit(X_mean_tr_s, yg_tr)
img_feat_tr_grd = svm_img_grd.decision_function(X_mean_tr_s)
img_feat_te_grd = svm_img_grd.decision_function(X_mean_te_s)
if img_feat_tr_grd.ndim == 1:
    img_feat_tr_grd = img_feat_tr_grd.reshape(-1, 1)
    img_feat_te_grd = img_feat_te_grd.reshape(-1, 1)

X_mid_grd_tr = np.hstack([img_feat_tr_grd, X_weight_tr])
X_mid_grd_te = np.hstack([img_feat_te_grd, X_weight_te])

scaler_mid_final_grd = RobustScaler()
X_mid_grd_tr_s = scaler_mid_final_grd.fit_transform(X_mid_grd_tr)
X_mid_grd_te_s = scaler_mid_final_grd.transform(X_mid_grd_te)

svm_mid_grd = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced', random_state=42)
svm_mid_grd.fit(X_mid_grd_tr_s, yg_tr)
yg_pred_mid = svm_mid_grd.predict(X_mid_grd_te_s)
yg_train_pred_mid = svm_mid_grd.predict(X_mid_grd_tr_s)

print(f"✓ Grade - F1={f1_score(yg_te, yg_pred_mid, average='weighted', zero_division=0):.4f}")

mid_results = pd.DataFrame([
    {
        'Fusion_Type': 'Mid-Level Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Ripeness',
        'Strategy': 'Two-Stage',
        'Train_Acc': round(accuracy_score(yr_tr, yr_train_pred_mid), 6),
        'Test_Acc': round(accuracy_score(yr_te, yr_pred_mid), 6),
        'Overfit_Gap': round(accuracy_score(yr_tr, yr_train_pred_mid) - accuracy_score(yr_te, yr_pred_mid), 6),
        'Accuracy': round(accuracy_score(yr_te, yr_pred_mid), 6),
        'Precision': round(precision_score(yr_te, yr_pred_mid, average='weighted', zero_division=0), 6),
        'Recall': round(recall_score(yr_te, yr_pred_mid, average='weighted', zero_division=0), 6),
        'F1_Score': round(f1_score(yr_te, yr_pred_mid, average='weighted', zero_division=0), 6)
    },
    {
        'Fusion_Type': 'Mid-Level Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Grade',
        'Strategy': 'Two-Stage',
        'Train_Acc': round(accuracy_score(yg_tr, yg_train_pred_mid), 6),
        'Test_Acc': round(accuracy_score(yg_te, yg_pred_mid), 6),
        'Overfit_Gap': round(accuracy_score(yg_tr, yg_train_pred_mid) - accuracy_score(yg_te, yg_pred_mid), 6),
        'Accuracy': round(accuracy_score(yg_te, yg_pred_mid), 6),
        'Precision': round(precision_score(yg_te, yg_pred_mid, average='weighted', zero_division=0), 6),
        'Recall': round(recall_score(yg_te, yg_pred_mid, average='weighted', zero_division=0), 6),
        'F1_Score': round(f1_score(yg_te, yg_pred_mid, average='weighted', zero_division=0), 6)
    }
])

mid_results.to_excel("svm_mid_fusion_results.xlsx", index=False)
print("\n✅ Mid-Level Fusion saved!")

# ======================================================
# FUSION 4: HYBRID
# ======================================================
print("\n" + "="*70)
print("FUSION 4: HYBRID FUSION")
print("="*70)

# Get probabilities from all three
scaler_early_rip = RobustScaler()
X_early_rip_tr_s = scaler_early_rip.fit_transform(X_early_rip_tr)
X_early_rip_te_s = scaler_early_rip.transform(X_early_rip_te)

svm_early_rip = SVC(kernel='rbf', C=best_rip_params['C'], gamma=best_rip_params['gamma'],
                    class_weight='balanced', random_state=42, probability=True)
svm_early_rip.fit(X_early_rip_tr_s, yr_tr)
early_rip_proba = svm_early_rip.predict_proba(X_early_rip_te_s)

svm_mid_rip_prob = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced',
                       random_state=42, probability=True)
svm_mid_rip_prob.fit(X_mid_rip_tr_s, yr_tr)
mid_rip_proba = svm_mid_rip_prob.predict_proba(X_mid_rip_te_s)

def late_proba(X_front, X_back, X_weight, y_train):
    probas = []
    for X in [X_front, X_back, X_weight]:
        scaler = RobustScaler()
        X_tr_s = scaler.fit_transform(X[0])
        X_te_s = scaler.transform(X[1])
        svm = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced',
                  random_state=42, probability=True)
        svm.fit(X_tr_s, y_train)
        probas.append(svm.predict_proba(X_te_s))
    return np.mean(probas, axis=0)

late_rip_proba = late_proba((X_front_tr, X_front_te), (X_back_tr, X_back_te),
                            (X_weight_tr, X_weight_te), yr_tr)

hybrid_rip_proba = (early_rip_proba + mid_rip_proba + late_rip_proba) / 3
hybrid_rip_pred = np.argmax(hybrid_rip_proba, axis=1)

print(f"✓ Ripeness - F1={f1_score(yr_te, hybrid_rip_pred, average='weighted', zero_division=0):.4f}")

# Grade
scaler_early_grd = RobustScaler()
X_early_grd_tr_s = scaler_early_grd.fit_transform(X_early_grd_tr)
X_early_grd_te_s = scaler_early_grd.transform(X_early_grd_te)

svm_early_grd = SVC(kernel='rbf', C=best_grd_params['C'], gamma=best_grd_params['gamma'],
                    class_weight='balanced', random_state=42, probability=True)
svm_early_grd.fit(X_early_grd_tr_s, yg_tr)
early_grd_proba = svm_early_grd.predict_proba(X_early_grd_te_s)

svm_mid_grd_prob = SVC(kernel='rbf', C=0.1, gamma='scale', class_weight='balanced',
                       random_state=42, probability=True)
svm_mid_grd_prob.fit(X_mid_grd_tr_s, yg_tr)
mid_grd_proba = svm_mid_grd_prob.predict_proba(X_mid_grd_te_s)

late_grd_proba = late_proba((X_front_tr, X_front_te), (X_back_tr, X_back_te),
                            (X_weight_tr, X_weight_te), yg_tr)

hybrid_grd_proba = (early_grd_proba + mid_grd_proba + late_grd_proba) / 3
hybrid_grd_pred = np.argmax(hybrid_grd_proba, axis=1)

print(f"✓ Grade - F1={f1_score(yg_te, hybrid_grd_pred, average='weighted', zero_division=0):.4f}")

hybrid_results = pd.DataFrame([
    {
        'Fusion_Type': 'Hybrid Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Ripeness',
        'Strategy': 'Ensemble (Early+Mid+Late)',
        'Accuracy': round(accuracy_score(yr_te, hybrid_rip_pred), 6),
        'Precision': round(precision_score(yr_te, hybrid_rip_pred, average='weighted', zero_division=0), 6),
        'Recall': round(recall_score(yr_te, hybrid_rip_pred, average='weighted', zero_division=0), 6),
        'F1_Score': round(f1_score(yr_te, hybrid_rip_pred, average='weighted', zero_division=0), 6)
    },
    {
        'Fusion_Type': 'Hybrid Fusion',
        'Model': 'SVM_RBF',
        'Task': 'Grade',
        'Strategy': 'Ensemble (Early+Mid+Late)',
        'Accuracy': round(accuracy_score(yg_te, hybrid_grd_pred), 6),
        'Precision': round(precision_score(yg_te, hybrid_grd_pred, average='weighted', zero_division=0), 6),
        'Recall': round(recall_score(yg_te, hybrid_grd_pred, average='weighted', zero_division=0), 6),
        'F1_Score': round(f1_score(yg_te, hybrid_grd_pred, average='weighted', zero_division=0), 6)
    }
])

hybrid_results.to_excel("svm_hybrid_fusion_results.xlsx", index=False)
print("\n✅ Hybrid Fusion saved!")

# ======================================================
# SUMMARY
# ======================================================
print("\n" + "="*70)
print("✅ SVM PIPELINE COMPLETE!")
print("="*70)
print("\nFiles Generated:")
print("  - svm_stage1_front.xlsx")
print("  - svm_stage2_back.xlsx")
print("  - svm_stage3_combined.xlsx")
print("  - svm_stage4_fused.xlsx")
print("  - svm_early_fusion_results.xlsx")
print("  - svm_late_fusion_results.xlsx")
print("  - svm_mid_fusion_results.xlsx")
print("  - svm_hybrid_fusion_results.xlsx")
```

```python
# ========================================
# MULTIMODAL MULTITASK
# GRADIENT BOOSTING & XGBOOST
# EARLY, LATE, HYBRID FUSION
# WITH DATA AUGMENTATION
# ========================================

import os
import cv2
import numpy as np
import pandas as pd
import warnings

from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load a sample dataset
data = load_iris()
X, y = data.data, data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Create and train model
model = XGBClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
from sklearn.base import clone # Import clone for model duplication

from tensorflow.keras.preprocessing.image import ImageDataGenerator

warnings.filterwarnings("ignore")

# ========================================
# PATHS
# ========================================

IMAGE_DIR = "images"
EXCEL_PATH = "Harumanis_mango_weight_grade.xlsx"

# ========================================
# LOAD DATA
# ========================================

df = pd.read_excel(EXCEL_PATH)
df["Fruit Grade"] = df["Fruit Grade"].astype(str)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

# ========================================
# IMAGE FEATURE EXTRACTION
# ========================================

def extract_image_features(img):
    img = cv2.resize(img, (224, 224))

    rgb_mean = img.mean(axis=(0,1))
    rgb_std = img.std(axis=(0,1))

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0,1))
    hsv_std = hsv.std(axis=(0,1))

    hue = hsv[:,:,0]
    hue_hist, _ = np.histogram(hue, bins=16, range=(0,180))
    hue_hist = hue_hist / hue_hist.sum()

    yellow = ((hue >= 20) & (hue <= 35)).sum() / hue.size
    green = ((hue >= 35) & (hue <= 85)).sum() / hue.size

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_r = (gray / 4).astype(np.uint8)

    glcm = graycomatrix(
        gray_r, [1],
        [0, np.pi/4, np.pi/2, 3*np.pi/4],
        levels=64, symmetric=True, normed=True
    )

    texture = [
        graycoprops(glcm, p).mean()
        for p in ["contrast", "homogeneity", "energy", "correlation", "dissimilarity"]
    ]

    edges = cv2.Canny(gray, 50, 150)
    edge_density = edges.sum() / edges.size

    return np.hstack([
        rgb_mean, rgb_std,
        hsv_mean, hsv_std,
        [yellow, green],
        hue_hist,
        texture,
        [edge_density]
    ])

# ========================================
# BUILD DATASET (RAW IMAGES + WEIGHT)
# ========================================

imgs, weights, y_rip, y_grd, groups = [], [], [], [], []

for fid, g in df.groupby("Fruit_ID"):
    if len(g) != 2:
        continue

    r1, r2 = g.iloc[0], g.iloc[1]

    img1 = cv2.imread(os.path.join(IMAGE_DIR, r1["Fruit No"]))
    img2 = cv2.imread(os.path.join(IMAGE_DIR, r2["Fruit No"]))

    if img1 is None or img2 is None:
        continue

    # Resize images to a common dimension before adding them
    # The extract_image_features function resizes to (224, 224), so use that here too.
    img1 = cv2.resize(img1, (224, 224))
    img2 = cv2.resize(img2, (224, 224))

    imgs.append((img1 + img2) / 2)
    weights.append([r1["Mass(kg)"]])
    y_rip.append(r1["Color_K-Yellow_P_Green"])
    y_grd.append(r1["Fruit Grade"])
    groups.append(fid)

imgs = np.array(imgs, dtype=np.float32)
weights = np.array(weights)
groups = np.array(groups)

# ========================================
# ENCODE LABELS
# ========================================

rip_enc = LabelEncoder()
grd_enc = LabelEncoder()

y_rip = rip_enc.fit_transform(y_rip)
y_grd = grd_enc.fit_transform(y_grd)

# ========================================
# GROUP-AWARE SPLIT
# ========================================

gss = GroupShuffleSplit(test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(imgs, y_rip, groups))

imgs_tr, imgs_te = imgs[train_idx], imgs[test_idx]
wt_tr, wt_te = weights[train_idx], weights[test_idx]

yr_tr, yr_te = y_rip[train_idx], y_rip[test_idx]
yg_tr, yg_te = y_grd[train_idx], y_grd[test_idx]

# ========================================
# DATA AUGMENTATION (TRAIN ONLY)
# ========================================

augmenter = ImageDataGenerator(
    rotation_range=15,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2]
)

aug_imgs = []
aug_yr = []
aug_yg = []
aug_wt = []

for img, yr, yg, wt in zip(imgs_tr, yr_tr, yg_tr, wt_tr):
    img = img.reshape(1, *img.shape)
    aug_iter = augmenter.flow(img, batch_size=1)

    for _ in range(2):  # 2 augmented samples
        aug_imgs.append(next(aug_iter)[0])
        aug_yr.append(yr)
        aug_yg.append(yg)
        aug_wt.append(wt)

imgs_tr_aug = np.vstack([imgs_tr, np.array(aug_imgs)])
yr_tr_aug = np.hstack([yr_tr, np.array(aug_yr)])
yg_tr_aug = np.hstack([yg_tr, np.array(aug_yg)])
wt_tr_aug = np.vstack([wt_tr, np.array(aug_wt)])

# ========================================
# FEATURE EXTRACTION
# ========================================

Xi_tr = np.array([extract_image_features(i.astype(np.uint8)) for i in imgs_tr_aug])
Xi_te = np.array([extract_image_features(i.astype(np.uint8)) for i in imgs_te])

# ========================================
# SCALING
# ========================================

sc_img = RobustScaler()
sc_wt = RobustScaler()

Xi_tr = sc_img.fit_transform(Xi_tr)
Xi_te = sc_img.transform(Xi_te)

wt_tr_aug = sc_wt.fit_transform(wt_tr_aug)
wt_te = sc_wt.transform(wt_te)

# ========================================
# FUSION DATASETS
# ========================================

X_early_tr = np.hstack([Xi_tr, wt_tr_aug])
X_early_te = np.hstack([Xi_te, wt_te])

# ========================================
# MODELS (Removed direct model instances here)
# ========================================

# Model names for iteration
model_types = ["Gradient Boosting", "XGBoost"]

# ========================================
# FUSION FUNCTIONS
# ========================================

def late_fusion(p_img, p_wt):
    return (p_img + p_wt) / 2

# ========================================
# TRAIN & EVALUATE
# ========================================

results = []

for model_name in model_types:

    if model_name == "Gradient Boosting":
        base_model_r = GradientBoostingClassifier(random_state=42)
        base_model_g = GradientBoostingClassifier(random_state=42)
    elif model_name == "XGBoost":
        # Determine num_class for ripeness (binary) and grade (multi-class)
        num_classes_ripeness = len(np.unique(yr_tr_aug)) # Should be 2
        num_classes_grade = len(np.unique(yg_tr_aug)) # Should be 3

        # For Ripeness (binary classification)
        base_model_r = XGBClassifier(
            random_state=42,
            eval_metric="logloss",
            use_label_encoder=False,
            objective='binary:logistic'
        )
        # For Grade (multi-class classification)
        base_model_g = XGBClassifier(
            random_state=42,
            eval_metric="mlogloss", # Multiclass logloss
            use_label_encoder=False,
            objective='multi:softmax',
            num_class=num_classes_grade
        )

    # Clone the base models for each fusion type and task to ensure fresh states
    # Early Fusion Models
    model_early_r = clone(base_model_r)
    model_early_g = clone(base_model_g)

    # Image Only Models
    model_img_r = clone(base_model_r)
    model_img_g = clone(base_model_g)

    # Weight Only Models
    model_wt_r = clone(base_model_r)
    model_wt_g = clone(base_model_g)

    # EARLY FUSION
    model_early_r.fit(X_early_tr, yr_tr_aug)
    p_early_r = model_early_r.predict_proba(X_early_te)

    model_early_g.fit(X_early_tr, yg_tr_aug)
    p_early_g = model_early_g.predict_proba(X_early_te)

    # IMAGE ONLY
    model_img_r.fit(Xi_tr, yr_tr_aug)
    p_img_r = model_img_r.predict_proba(Xi_te)

    model_img_g.fit(Xi_tr, yg_tr_aug)
    p_img_g = model_img_g.predict_proba(Xi_te)

    # WEIGHT ONLY
    model_wt_r.fit(wt_tr_aug, yr_tr_aug)
    p_wt_r = model_wt_r.predict_proba(wt_te)

    model_wt_g.fit(wt_tr_aug, yg_tr_aug)
    p_wt_g = model_wt_g.predict_proba(wt_te)

    # LATE FUSION
    p_late_r = late_fusion(p_img_r, p_wt_r)
    p_late_g = late_fusion(p_img_g, p_wt_g)

    # HYBRID FUSION
    p_hybrid_r = (p_early_r + p_late_r) / 2
    p_hybrid_g = (p_early_g + p_late_g) / 2

    fusion_sets = {
        "Early Fusion": (p_early_r, p_early_g),
        "Late Fusion": (p_late_r, p_late_g),
        "Hybrid Fusion": (p_hybrid_r, p_hybrid_g)
    }

    for fusion, (pr, pg) in fusion_sets.items():
        results.append({
            "Model": model_name,
            "Fusion": fusion,
            "Task": "Ripeness",
            "Accuracy": accuracy_score(yr_te, np.argmax(pr, axis=1)),
            "Precision": precision_score(yr_te, np.argmax(pr, axis=1), average="weighted", zero_division=0),
            "Recall": recall_score(yr_te, np.argmax(pr, axis=1), average="weighted", zero_division=0),
            "F1": f1_score(yr_te, np.argmax(pr, axis=1), average="weighted", zero_division=0)
        })
        results.append({
            "Model": model_name,
            "Fusion": fusion,
            "Task": "Grade",
            "Accuracy": accuracy_score(yg_te, np.argmax(pg, axis=1)),
            "Precision": precision_score(yg_te, np.argmax(pg, axis=1), average="weighted", zero_division=0),
            "Recall": recall_score(yg_te, np.argmax(pg, axis=1), average="weighted", zero_division=0),
            "F1": f1_score(yg_te, np.argmax(pg, axis=1), average="weighted", zero_division=0)
        })


results_df = pd.DataFrame(results)

print("\n=== FUSION MODEL RESULTS ===")
print(results_df.round(3).to_string(index=False))

results_df.to_excel("fusion_model_results.xlsx", index=False)

print("\nResults saved to: fusion_model_results.xlsx")
print("Execution completed successfully.")
```

# Rnadom Forest And Naive Bayes

---
# 🟠 PHASE 2: Multimodal Fusion — Early | Late | Mid | Hybrid
## Naive Bayes & Random Forest
**Targets:** Ripeness + Grade

| Fusion | Strategy |
|--------|----------|
| **Early** | Concatenate image + weight **before** training — one model sees everything |
| **Late** | Train separate model per modality, **average probabilities after** prediction |
| **Mid** | Weighted feature blend at input level — image × 0.7 + weight × 0.3 |
| **Hybrid** | Average of Early + Late + Mid probabilities |

### 🏋️ Train & Evaluate — Phase 2

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.base import clone

def get_metrics(model_name, fusion_name, task, y_true, y_pred):
    return {
        "Model":     model_name,
        "Fusion":    fusion_name,
        "Task":      task,
        "Accuracy":  accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average="weighted", zero_division=0),
        "Recall":    recall_score(y_true, y_pred, average="weighted", zero_division=0),
        "F1":        f1_score(y_true, y_pred, average="weighted", zero_division=0)
    }

model_configs = {
    "Naive Bayes":   GaussianNB(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

results_p2 = []

for model_name, base_model in model_configs.items():

    print(f"\n{'='*55}")
    print(f"  Model: {model_name}")
    print(f"{'='*55}")

    m_early_r = clone(base_model);  m_img_r = clone(base_model)
    m_early_g = clone(base_model);  m_img_g = clone(base_model)
    m_wt_r    = clone(base_model);  m_mid_r = clone(base_model)
    m_wt_g    = clone(base_model);  m_mid_g = clone(base_model)

    # ── 1️⃣ EARLY FUSION ─────────────────────────────────────────────────
    print("  ▶ Early Fusion  (image + weight concatenated before training)...")
    m_early_r.fit(X_early_tr, yr_tr_aug);  p_early_r = m_early_r.predict_proba(X_early_te)
    m_early_g.fit(X_early_tr, yg_tr_aug);  p_early_g = m_early_g.predict_proba(X_early_te)

    # ── Modality-specific models ──────────────────────────────────────────
    m_img_r.fit(Xi_tr, yr_tr_aug);    p_img_r = m_img_r.predict_proba(Xi_te)
    m_img_g.fit(Xi_tr, yg_tr_aug);    p_img_g = m_img_g.predict_proba(Xi_te)
    m_wt_r.fit(wt_tr_aug, yr_tr_aug); p_wt_r  = m_wt_r.predict_proba(wt_te)
    m_wt_g.fit(wt_tr_aug, yg_tr_aug); p_wt_g  = m_wt_g.predict_proba(wt_te)

    # ── 2️⃣ LATE FUSION ───────────────────────────────────────────────────
    print("  ▶ Late Fusion   (average probabilities after separate training)...")
    p_late_r = (p_img_r + p_wt_r) / 2
    p_late_g = (p_img_g + p_wt_g) / 2

    # ── 3️⃣ MID FUSION ────────────────────────────────────────────────────
    print("  ▶ Mid Fusion    (weighted feature blend: image×0.7 + weight×0.3)...")
    X_mid_tr = np.hstack([Xi_tr * 0.7, wt_tr_aug * 0.3])
    X_mid_te = np.hstack([Xi_te * 0.7, wt_te     * 0.3])
    m_mid_r.fit(X_mid_tr, yr_tr_aug);  p_mid_r = m_mid_r.predict_proba(X_mid_te)
    m_mid_g.fit(X_mid_tr, yg_tr_aug);  p_mid_g = m_mid_g.predict_proba(X_mid_te)

    # ── 4️⃣ HYBRID FUSION ─────────────────────────────────────────────────
    print("  ▶ Hybrid Fusion (average of Early + Late + Mid)...")
    p_hybrid_r = (p_early_r + p_late_r + p_mid_r) / 3
    p_hybrid_g = (p_early_g + p_late_g + p_mid_g) / 3

    fusion_sets = {
        "Early Fusion":  (p_early_r,  p_early_g),
        "Late Fusion":   (p_late_r,   p_late_g),
        "Mid Fusion":    (p_mid_r,    p_mid_g),
        "Hybrid Fusion": (p_hybrid_r, p_hybrid_g),
    }

    print()
    for fusion_name, (pr, pg) in fusion_sets.items():
        pred_r = np.argmax(pr, axis=1)
        pred_g = np.argmax(pg, axis=1)
        results_p2.append(get_metrics(model_name, fusion_name, "Ripeness", yr_te, pred_r))
        results_p2.append(get_metrics(model_name, fusion_name, "Grade",    yg_te, pred_g))
        print(f"    [{fusion_name:<14}]  Ripeness Acc: {accuracy_score(yr_te, pred_r):.3f} | "
              f"Grade Acc: {accuracy_score(yg_te, pred_g):.3f}")

print("\n✅ Phase 2 complete!")
```

```python
results_p2_df = pd.DataFrame(results_p2)

print("=== PHASE 2: FUSION MODEL RESULTS ===")
print(results_p2_df.round(3).to_string(index=False))

results_p2_df.to_excel("phase2_fusion_results.xlsx", index=False)
print("\n✅ Saved to: phase2_fusion_results.xlsx")
```

# KNN And Decision Tree

```python

import numpy as np
import pandas as pd

from sklearn.model_selection import GroupShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python

# ===============================
# REPLACE THIS SECTION WITH YOUR REAL DATA
# ===============================

np.random.seed(42)
X = np.random.rand(200, 21)   # 20 image features + 1 weight
y_r = np.random.randint(0,2,200)   # Ripeness
y_g = np.random.randint(0,3,200)   # Grade
groups = np.random.randint(0,50,200)
```

```python

gss = GroupShuffleSplit(test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y_r, groups))

X_train, X_test = X[train_idx], X[test_idx]
yr_train, yr_test = y_r[train_idx], y_r[test_idx]
yg_train, yg_test = y_g[train_idx], y_g[test_idx]

X_img = X[:, :-1]
X_w = X[:, -1:].reshape(-1,1)

X_img_train, X_img_test = X_img[train_idx], X_img[test_idx]
X_w_train, X_w_test = X_w[train_idx], X_w[test_idx]
```

```python

def compute_metrics(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_true, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(y_true, y_pred, average="weighted", zero_division=0)
    }
```

```python

results = []

# EARLY FUSION
knn = KNeighborsClassifier(5)
dt = DecisionTreeClassifier(max_depth=5, random_state=42)

# KNN Early
knn.fit(X_train, yr_train)
results.append({"Model":"KNN","Fusion":"Early","Task":"Ripeness", **compute_metrics(yr_test, knn.predict(X_test))})
knn.fit(X_train, yg_train)
results.append({"Model":"KNN","Fusion":"Early","Task":"Grade", **compute_metrics(yg_test, knn.predict(X_test))})

# DT Early
dt.fit(X_train, yr_train)
results.append({"Model":"Decision Tree","Fusion":"Early","Task":"Ripeness", **compute_metrics(yr_test, dt.predict(X_test))})
dt.fit(X_train, yg_train)
results.append({"Model":"Decision Tree","Fusion":"Early","Task":"Grade", **compute_metrics(yg_test, dt.predict(X_test))})

# LATE FUSION
def late(model, X_img_tr, X_w_tr, X_img_te, X_w_te, y_tr):
    m1 = model()
    m2 = model()
    m1.fit(X_img_tr, y_tr)
    m2.fit(X_w_tr, y_tr)
    prob = (m1.predict_proba(X_img_te) + m2.predict_proba(X_w_te))/2
    return np.argmax(prob, axis=1)

# KNN Late
results.append({"Model":"KNN","Fusion":"Late","Task":"Ripeness",
                **compute_metrics(yr_test, late(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"KNN","Fusion":"Late","Task":"Grade",
                **compute_metrics(yg_test, late(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# DT Late
results.append({"Model":"Decision Tree","Fusion":"Late","Task":"Ripeness",
                **compute_metrics(yr_test, late(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Decision Tree","Fusion":"Late","Task":"Grade",
                **compute_metrics(yg_test, late(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# HYBRID FUSION
def hybrid(model, X_img_tr, X_w_tr, X_img_te, X_w_te, y_tr):
    m1 = model()
    m2 = model()
    m1.fit(X_img_tr, y_tr)
    m2.fit(X_w_tr, y_tr)
    prob = 0.6*m1.predict_proba(X_img_te) + 0.4*m2.predict_proba(X_w_te)
    return np.argmax(prob, axis=1)

# KNN Hybrid
results.append({"Model":"KNN","Fusion":"Hybrid","Task":"Ripeness",
                **compute_metrics(yr_test, hybrid(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"KNN","Fusion":"Hybrid","Task":"Grade",
                **compute_metrics(yg_test, hybrid(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# DT Hybrid
results.append({"Model":"Decision Tree","Fusion":"Hybrid","Task":"Ripeness",
                **compute_metrics(yr_test, hybrid(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Decision Tree","Fusion":"Hybrid","Task":"Grade",
                **compute_metrics(yg_test, hybrid(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})
```

```python

results_df = pd.DataFrame(results).round(4)
print(results_df)
results_df.to_excel("final_fusion_results.xlsx", index=False)
print("Saved as final_fusion_results.xlsx")
```

```python
import pandas as pd

# 1. Convert global_results list into a single DataFrame
final_df = pd.DataFrame(global_results)

# 2. Standardize 'Fusion Strategy' names
# Rename 'Fusion_Type' to 'Fusion' if it exists, then consolidate
if 'Fusion_Type' in final_df.columns:
    final_df['Fusion'] = final_df['Fusion'].fillna(final_df['Fusion_Type'])
    final_df = final_df.drop(columns=['Fusion_Type'])

fusion_mapping = {
    'Early': 'Early Fusion', 'Early Fusion': 'Early Fusion',
    'Late': 'Late Fusion', 'Late Fusion': 'Late Fusion',
    'Mid': 'Mid-Level Fusion', 'Mid Fusion': 'Mid-Level Fusion', 'Mid-Level Fusion': 'Mid-Level Fusion',
    'Hybrid': 'Hybrid Fusion', 'Hybrid Fusion': 'Hybrid Fusion'
}
final_df['Fusion Strategy'] = final_df['Fusion'].map(fusion_mapping)

# 3. Consolidate F1 metric columns
if 'F1_Score' in final_df.columns:
    final_df['F1'] = final_df['F1'].fillna(final_df['F1_Score'])
    final_df = final_df.drop(columns=['F1_Score']) # Drop the old F1_Score column

# Consolidate Accuracy, Precision, Recall as well
if 'Accuracy_Score' in final_df.columns:
    final_df['Accuracy'] = final_df['Accuracy'].fillna(final_df['Accuracy_Score'])
    final_df = final_df.drop(columns=['Accuracy_Score'])
if 'Precision_Score' in final_df.columns:
    final_df['Precision'] = final_df['Precision'].fillna(final_df['Precision_Score'])
    final_df = final_df.drop(columns=['Precision_Score'])
if 'Recall_Score' in final_df.columns:
    final_df['Recall'] = final_df['Recall'].fillna(final_df['Recall_Score'])
    final_df = final_df.drop(columns=['Recall_Score'])


# Drop any redundant columns that might have come from different schemas
# For example, from SVM results, we might have 'Best_C', 'Best_Gamma', 'Train_Acc', 'Test_Acc', 'Overfit_Gap', 'Strategy'
columns_to_drop_if_exist = ['Best_C', 'Best_Gamma', 'Train_Acc', 'Test_Acc', 'Overfit_Gap', 'Strategy', 'Fusion']
final_df = final_df.drop(columns=[col for col in columns_to_drop_if_exist if col in final_df.columns])

# 4. Remove duplicate rows based on unique experiment identifiers
final_df = final_df.drop_duplicates(subset=['Model', 'Task', 'Fusion Strategy']).reset_index(drop=True)

# 5. Add 'Phase' column back at the first position
final_df.insert(0, 'Phase', 'Phase 3')

# 6. Apply custom categorical sort for Model order
model_order = [
    'Gradient Boosting', 'SVM_RBF', 'Logistic Regression', 'XGBoost',
    'Naive Bayes', 'Random Forest', 'KNN', 'Decision Tree'
]
final_df['Model'] = pd.Categorical(final_df['Model'], categories=model_order, ordered=True)

# Apply custom categorical sort for Task order: Ripeness then Grade
task_order = ['Ripeness', 'Grade']
final_df['Task'] = pd.Categorical(final_df['Task'], categories=task_order, ordered=True)

# Sort by Task (Ripeness first), then F1 (descending), then Model (custom order)
final_df = final_df.sort_values(by=['Task', 'F1', 'Model'], ascending=[True, False, True])

# 7. Round numeric metrics to 4 decimal places
metric_cols = ['Accuracy', 'Precision', 'Recall', 'F1']
final_df[metric_cols] = final_df[metric_cols].apply(pd.to_numeric, errors='coerce').round(4)

# Select final columns for output
final_columns = ['Phase', 'Model', 'Task', 'Fusion Strategy', 'Accuracy', 'Precision', 'Recall', 'F1']
final_df = final_df[final_columns]

# 8. Export and Display without the index numbers
final_df.to_excel('phase3_results_final.xlsx', index=False)
print('=== FINAL PHASE 3 CONSOLIDATED RESULTS ===')
print(f'Total records exported: {len(final_df)}')
# Display the dataframe as a string or hide the index to remove the numbering column
from IPython.display import display
display(final_df.style.hide(axis='index'))
```

```

import numpy as np
import pandas as pd

from sklearn.model_selection import GroupShuffleSplit
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```

# ===============================
# REPLACE THIS SECTION WITH YOUR REAL DATA
# ===============================

np.random.seed(42)
X = np.random.rand(200, 21)   # 20 image features + 1 weight
y_r = np.random.randint(0,2,200)   # Ripeness
y_g = np.random.randint(0,3,200)   # Grade
groups = np.random.randint(0,50,200)
```

```

gss = GroupShuffleSplit(test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y_r, groups))

X_train, X_test = X[train_idx], X[test_idx]
yr_train, yr_test = y_r[train_idx], y_r[test_idx]
yg_train, yg_test = y_g[train_idx], y_g[test_idx]

# Separate image & weight
X_img = X[:, :-1]
X_w = X[:, -1:].reshape(-1,1)

X_img_train, X_img_test = X_img[train_idx], X_img[test_idx]
X_w_train, X_w_test = X_w[train_idx], X_w[test_idx]
```

```

def compute_metrics(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_true, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(y_true, y_pred, average="weighted", zero_division=0)
    }
```

```

results = []

# =========================
# EARLY FUSION
# =========================

nb_model = GaussianNB()
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Naive Bayes Early
nb_model.fit(X_train, yr_train)
results.append({"Model":"Naive Bayes","Fusion":"Early","Task":"Ripeness", **compute_metrics(yr_test, nb_model.predict(X_test))})
nb_model.fit(X_train, yg_train)
results.append({"Model":"Naive Bayes","Fusion":"Early","Task":"Grade", **compute_metrics(yg_test, nb_model.predict(X_test))})

# Random Forest Early
rf_model.fit(X_train, yr_train)
results.append({"Model":"Random Forest","Fusion":"Early","Task":"Ripeness", **compute_metrics(yr_test, rf_model.predict(X_test))})
rf_model.fit(X_train, yg_train)
results.append({"Model":"Random Forest","Fusion":"Early","Task":"Grade", **compute_metrics(yg_test, rf_model.predict(X_test))})

# =========================
# LATE FUSION
# =========================

def late(model, X_img_tr, X_w_tr, X_img_te, X_w_te, y_tr):
    m1 = model()
    m2 = model()
    m1.fit(X_img_tr, y_tr)
    m2.fit(X_w_tr, y_tr)
    prob = (m1.predict_proba(X_img_te) + m2.predict_proba(X_w_te)) / 2
    return np.argmax(prob, axis=1)

# NB Late
results.append({"Model":"Naive Bayes","Fusion":"Late","Task":"Ripeness",
                **compute_metrics(yr_test, late(GaussianNB,
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Naive Bayes","Fusion":"Late","Task":"Grade",
                **compute_metrics(yg_test, late(GaussianNB,
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# RF Late
results.append({"Model":"Random Forest","Fusion":"Late","Task":"Ripeness",
                **compute_metrics(yr_test, late(lambda:RandomForestClassifier(n_estimators=100,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Random Forest","Fusion":"Late","Task":"Grade",
                **compute_metrics(yg_test, late(lambda:RandomForestClassifier(n_estimators=100,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# =========================
# HYBRID FUSION (0.6 Image + 0.4 Weight)
# =========================

def hybrid(model, X_img_tr, X_w_tr, X_img_te, X_w_te, y_tr):
    m1 = model()
    m2 = model()
    m1.fit(X_img_tr, y_tr)
    m2.fit(X_w_tr, y_tr)
    prob = 0.6*m1.predict_proba(X_img_te) + 0.4*m2.predict_proba(X_w_te)
    return np.argmax(prob, axis=1)

# NB Hybrid
results.append({"Model":"Naive Bayes","Fusion":"Hybrid","Task":"Ripeness",
                **compute_metrics(yr_test, hybrid(GaussianNB,
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Naive Bayes","Fusion":"Hybrid","Task":"Grade",
                **compute_metrics(yg_test, hybrid(GaussianNB,
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# RF Hybrid
results.append({"Model":"Random Forest","Fusion":"Hybrid","Task":"Ripeness",
                **compute_metrics(yr_test, hybrid(lambda:RandomForestClassifier(n_estimators=100,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Random Forest","Fusion":"Hybrid","Task":"Grade",
                **compute_metrics(yg_test, hybrid(lambda:RandomForestClassifier(n_estimators=100,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})
```

```

results_df = pd.DataFrame(results).round(4)
print(results_df)

results_df.to_excel("nb_rf_fusion_results.xlsx", index=False)
print("Saved as nb_rf_fusion_results.xlsx")
```

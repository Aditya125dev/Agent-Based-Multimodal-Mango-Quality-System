# KNN and Decision Tree Training on Image Dataset



```python
import os
import numpy as np
import pandas as pd
from PIL import Image

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
EXCEL_PATH = "Maturity_dataset.xlsx"
IMAGE_FOLDER = "images"

IMAGE_COLUMN = "Fruit No"        # image filename column
TARGET_COLUMN = "Maturity"       # class / grade / maturity label
```

```python
df = pd.read_excel("Maturity_dataset.xlsx")

print(df.head())
print(df.columns.tolist())
```

```python

# Image preprocessing function
def load_image(path, size=(64, 64)):
    img = Image.open(path).convert("RGB")
    img = img.resize(size)
    return np.array(img).flatten()
```

```python
X = []
y = []
groups = []

for _, row in df.iterrows():
    img_name = str(row[IMAGE_COLUMN]).strip()
    img_path = os.path.join(IMAGE_FOLDER, img_name)

    if os.path.exists(img_path):
        X.append(load_image(img_path))
        y.append(row[TARGET_COLUMN])
        
        # extract numeric fruit ID (1a.jpg → 1)
        group_id = "".join(filter(str.isdigit, img_name))
        groups.append(group_id)
    else:
        print("Missing image:", img_path)

X = np.array(X)
y = np.array(y)
groups = np.array(groups)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("groups shape:", groups.shape)
```

```python
from sklearn.model_selection import GroupShuffleSplit

gss = GroupShuffleSplit(test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups))

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]
```

```python
results = []

# -------- KNN --------
k_values = [3, 5, 7, 9]

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results.append({
        "Model": "KNN",
        "Modification": f"k = {k}",
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(y_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_test, y_pred, average="weighted", zero_division=0)
    })


# -------- DECISION TREE --------
depths = [3, 5, 10, None]

for d in depths:
    model = DecisionTreeClassifier(max_depth=d, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results.append({
        "Model": "Decision Tree",
        "Modification": f"max_depth = {d}",
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(y_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_test, y_pred, average="weighted", zero_division=0)
    })
```

```python

results_df = pd.DataFrame(results)

results_df[["Accuracy", "Precision", "F1", "Recall"]] = (
    results_df[["Accuracy", "Precision", "F1", "Recall"]].round(4)
)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
results_df
```

```python
import time
import matplotlib.pyplot as plt
timing_results = []

# ---------- KNN ----------
k_values = [3, 5, 7, 9]

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)

    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()

    timing_results.append({
        "Model": "KNN",
        "Modification": f"k = {k}",
        "Training_Time": end - start
    })


# ---------- Decision Tree ----------
depths = [3, 5, 10, None]

for d in depths:
    model = DecisionTreeClassifier(max_depth=d, random_state=42)

    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()

    timing_results.append({
        "Model": "Decision Tree",
        "Modification": f"max_depth = {d}",
        "Training_Time": end - start
    })

timing_df = pd.DataFrame(timing_results)
timing_df
```

```python
avg_times = timing_df.groupby("Model")["Training_Time"].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_times.index, avg_times.values)
plt.ylabel("Average Training Time (seconds)")
plt.title("Training Time Comparison (KNN vs Decision Tree)")
plt.tight_layout()
plt.savefig("training_time_comparison.png", dpi=300)
plt.show()
```

```python
# ===============================
# FUSION COMPARISON: EARLY vs LATE vs HYBRID
# Assumes X, train_idx, test_idx, X_train_s, X_test_s, yr_train, yr_test exist
# ===============================

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd

# -------- EARLY FUSION --------
lr_early = LogisticRegression(max_iter=2000, class_weight='balanced', C=0.1)
lr_early.fit(X_train_s, yr_train)
yr_pred_early = lr_early.predict(X_test_s)

# -------- LATE FUSION (image vs weight) --------
X_img = X[:, :-1]
X_weight = X[:, -1:].reshape(-1,1)

X_img_train, X_img_test = X_img[train_idx], X_img[test_idx]
X_w_train, X_w_test = X_weight[train_idx], X_weight[test_idx]

sc_img = RobustScaler()
sc_w = RobustScaler()

X_img_train_s = sc_img.fit_transform(X_img_train)
X_img_test_s = sc_img.transform(X_img_test)

X_w_train_s = sc_w.fit_transform(X_w_train)
X_w_test_s = sc_w.transform(X_w_test)

lr_img = LogisticRegression(max_iter=2000, class_weight='balanced')
lr_w = LogisticRegression(max_iter=2000, class_weight='balanced')

lr_img.fit(X_img_train_s, yr_train)
lr_w.fit(X_w_train_s, yr_train)

p_img = lr_img.predict_proba(X_img_test_s)
p_w = lr_w.predict_proba(X_w_test_s)

late_prob = (p_img + p_w) / 2
yr_pred_late = np.argmax(late_prob, axis=1)

# -------- HYBRID FUSION --------
p_comb = lr_early.predict_proba(X_test_s)
hybrid_prob = 0.4*p_img + 0.3*p_w + 0.3*p_comb
yr_pred_hybrid = np.argmax(hybrid_prob, axis=1)

# -------- TABULATE RESULTS --------
def get_metrics(y_true, y_hat):
    return {
        "Accuracy": accuracy_score(y_true, y_hat),
        "Precision": precision_score(y_true, y_hat, average="weighted", zero_division=0),
        "Recall": recall_score(y_true, y_hat, average="weighted", zero_division=0),
        "F1": f1_score(y_true, y_hat, average="weighted", zero_division=0)
    }

results = [
    {"Fusion": "Early", **get_metrics(yr_test, yr_pred_early)},
    {"Fusion": "Late", **get_metrics(yr_test, yr_pred_late)},
    {"Fusion": "Hybrid", **get_metrics(yr_test, yr_pred_hybrid)}
]

fusion_results_df = pd.DataFrame(results).round(4)
print(fusion_results_df)

fusion_results_df.to_excel("fusion_comparison_results.xlsx", index=False)
print("Saved: fusion_comparison_results.xlsx")
```

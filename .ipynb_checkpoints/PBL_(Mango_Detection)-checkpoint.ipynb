```python
# ==========================================================================================================
# ADVANCED AGENTIC AI + DECISION SUPPORT SYSTEM FOR HARUMANIS MANGO GRADING
# ==========================================================================================================
#
# CRITICAL FIX: Mango detection optimized for studio/white background images
#
# Author: AI Engineer
# Date: 2026-03-21
# Version: 2.1 (FIXED)
#
# ==========================================================================================================

import os
import cv2
import numpy as np
import pandas as pd
import warnings
from typing import Dict, Tuple, Optional

from skimage.feature import graycomatrix, graycoprops
from skimage.metrics import structural_similarity as ssim

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import RobustScaler, LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import GradientBoostingClassifier

try:
    import tensorflow as tf
    from tensorflow.keras.applications import MobileNetV2
    from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
    DEEP_LEARNING_AVAILABLE = True
except ImportError:
    DEEP_LEARNING_AVAILABLE = False
    print("⚠️  WARNING: TensorFlow not installed. Using color-based mango detection.")

warnings.filterwarnings("ignore")


# ==========================================================================================================
# GLOBAL CONFIGURATION
# ==========================================================================================================

class Config:
    """Centralized configuration"""

    IMAGE_DIR = "/content/images"
    EXCEL_PATH = "/content/Harumanis_mango_weight_grade.xlsx"

    class Validation:
        # CRITICAL FIX: Lowered thresholds for studio images
        MANGO_CONFIDENCE_THRESHOLD = 0.05      # Was 0.30, now VERY LENIENT
        MANGO_COLOR_PIXEL_RATIO = 0.10         # Was 0.30, now 0.10

        BLUR_THRESHOLD_BASE = 35
        BLUR_DARK_MULTIPLIER = 0.8
        BLUR_BRIGHT_MULTIPLIER = 1.2

        BRIGHTNESS_MIN = 40
        BRIGHTNESS_MAX = 220

        SSIM_DUPLICATE_THRESHOLD = 0.95

        HISTOGRAM_CORRELATION_MIN = 0.85
        SSIM_SAME_MANGO_MIN = 0.75
        ORB_MATCHES_MIN = 8

    class ML:
        CV_FOLDS = 5
        CV_RANDOM_STATE = 42

        GB_N_ESTIMATORS = 50
        GB_LEARNING_RATE = 0.05
        GB_MAX_DEPTH = 3
        GB_MIN_SAMPLES_SPLIT = 10
        GB_MIN_SAMPLES_LEAF = 5
        GB_SUBSAMPLE = 0.8
        GB_RANDOM_STATE = 42

    class DSS:
        class ShelfLife:
            MAX_DAYS = 14
            MIN_DAYS = 2
            WEIGHT_RIPENESS = 0.40
            WEIGHT_DEFECTS = 0.35
            WEIGHT_TEXTURE = 0.25

        class MarketRouting:
            EXPORT_THRESHOLD = 0.75
            RETAIL_THRESHOLD = 0.50
            WEIGHT_GRADE = 0.50
            WEIGHT_SHELF_LIFE = 0.30
            WEIGHT_WEIGHT = 0.20
            OPTIMAL_WEIGHT_MIN = 450
            OPTIMAL_WEIGHT_MAX = 600

    class Features:
        IMAGE_SIZE = (224, 224)
        HSV_LOWER = [15, 30, 30]
        HSV_UPPER = [95, 255, 255]
        GLCM_DISTANCES = [1]
        GLCM_ANGLES = [0, np.pi/4, np.pi/2, 3*np.pi/4]
        GLCM_LEVELS = 64
        DARK_SPOT_THRESHOLD = 50
        ABNORMAL_SAT_LOW = 20
        ABNORMAL_SAT_HIGH = 250


# ==========================================================================================================
# FIXED MANGO DETECTOR - OPTIMIZED FOR STUDIO IMAGES
# ==========================================================================================================

class MangoDetector:
    """
    IMPROVED: Balanced mango detection

    Strategy:
      - MobileNetV2: Accept "mango" specifically, or generic "fruit" with color verification
      - Color-based: Stricter green-to-yellow range check
      - Hybrid: Require BOTH methods to agree for non-mango fruits
    """

    def __init__(self):
        """Initialize with balanced thresholds"""

        # Specific mango detection
        self.mango_confidence_threshold = 0.10  # If MobileNetV2 says "mango", accept at 10%
        self.fruit_confidence_threshold = 0.30  # Generic "fruit" needs higher confidence + color check

        # Mango-specific classes
        self.mango_classes = ['mango']

        # Generic fruit classes (require additional verification)
        self.fruit_classes = ['fruit', 'orange', 'lemon', 'pomegranate', 'fig']

        # Non-mango fruits to reject
        self.reject_classes = ['apple', 'strawberry', 'banana', 'pineapple', 'grapes', 'cherry']

        if DEEP_LEARNING_AVAILABLE:
            try:
                self.model = MobileNetV2(weights='imagenet', include_top=True)
                self.use_deep_learning = True
                print("✓ MobileNetV2 loaded for mango detection")
            except Exception as e:
                print(f"⚠️  MobileNetV2 load failed: {e}")
                self.model = None
                self.use_deep_learning = False
        else:
            self.model = None
            self.use_deep_learning = False

    def detect_deep_learning(self, img: np.ndarray) -> Tuple[bool, float, str, str]:
        """
        Enhanced MobileNetV2 detection with fruit classification

        Returns:
            is_mango: Boolean
            confidence: Float
            detected_class: String (class name)
            detection_type: "specific_mango" | "generic_fruit" | "rejected" | "unknown"
        """

        img_resized = cv2.resize(img, Config.Features.IMAGE_SIZE)
        img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
        img_batch = np.expand_dims(img_rgb, axis=0)
        img_preprocessed = preprocess_input(img_batch)

        predictions = self.model.predict(img_preprocessed, verbose=0)
        decoded = decode_predictions(predictions, top=10)[0]

        # Check predictions in order
        for _, class_name, confidence in decoded:
            class_lower = class_name.lower()

            # PRIORITY 1: Explicit rejection (non-mango fruits)
            if any(reject in class_lower for reject in self.reject_classes):
                return False, float(confidence), class_name, "rejected_fruit"

            # PRIORITY 2: Specific mango detection
            if any(mango in class_lower for mango in self.mango_classes):
                if confidence >= self.mango_confidence_threshold:
                    return True, float(confidence), class_name, "specific_mango"

            # PRIORITY 3: Generic fruit (needs color verification)
            if any(fruit in class_lower for fruit in self.fruit_classes):
                if confidence >= self.fruit_confidence_threshold:
                    return True, float(confidence), class_name, "generic_fruit"

        # Not detected
        return False, 0.0, "unknown", "unknown"

    def detect_color_based(self, img: np.ndarray) -> Tuple[bool, float]:
        """
        STRICTER color-based detection for mango-specific colors

        Mango color range: Green to yellow (Hue 25-90 in HSV)
        Apple/Red fruits: Different hue range
        """

        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

        # TIGHTENED: More specific mango color range
        # Hue: 25-90 (green to yellow, excludes red apples at 0-15)
        # Sat: 40-255 (requires some color saturation)
        # Val: 40-255 (not too dark)
        lower = np.array([25, 40, 40])   # More restrictive than [15, 30, 30]
        upper = np.array([90, 255, 255]) # Tighter upper bound

        mask = cv2.inRange(hsv, lower, upper)

        mango_pixels = cv2.countNonZero(mask)
        total_pixels = img.shape[0] * img.shape[1]
        pixel_ratio = mango_pixels / total_pixels

        # STRICTER: Need at least 25% mango-colored pixels
        confidence = min(pixel_ratio * 3.0, 1.0)
        is_mango = pixel_ratio > 0.25  # Increased from 0.10

        return is_mango, confidence

    def detect(self, img: np.ndarray) -> Dict:
        """
        IMPROVED hybrid detection with smart logic

        Logic:
          1. Run MobileNetV2
          2. If "rejected_fruit" (apple, etc.) → REJECT immediately
          3. If "specific_mango" → ACCEPT (with color check for confidence boost)
          4. If "generic_fruit" → Require color verification
          5. If "unknown" → Use color-based only
        """

        ml_detected = False
        ml_confidence = 0.0
        ml_class = "unknown"
        detection_type = "unknown"

        # Run MobileNetV2
        if self.use_deep_learning:
            ml_detected, ml_confidence, ml_class, detection_type = self.detect_deep_learning(img)

        # Run color-based
        color_detected, color_confidence = self.detect_color_based(img)

        # DECISION LOGIC
        if detection_type == "rejected_fruit":
            # Explicitly rejected (apple, banana, etc.)
            is_mango = False
            final_confidence = ml_confidence
            method = f"Rejected: {ml_class}"

        elif detection_type == "specific_mango":
            # MobileNetV2 specifically detected mango
            is_mango = True
            final_confidence = max(ml_confidence, color_confidence)
            method = "MobileNetV2 (Mango)"

        elif detection_type == "generic_fruit":
            # Generic fruit - require color verification
            is_mango = ml_detected and color_detected
            final_confidence = (ml_confidence + color_confidence) / 2
            method = "Hybrid (Fruit + Color)" if is_mango else "Generic Fruit (Failed Color Check)"

        else:
            # Unknown or no ML - use color only
            is_mango = color_detected
            final_confidence = color_confidence
            method = "Color-Based Only"

        return {
            'is_mango': is_mango,
            'confidence': final_confidence,
            'detected_class': ml_class if ml_class != "unknown" else "color_based",
            'method': method,
            'detection_details': {
                'ml_detected': ml_detected,
                'ml_confidence': ml_confidence,
                'ml_class': ml_class,
                'detection_type': detection_type,
                'color_detected': color_detected,
                'color_confidence': color_confidence
            }
        }
# ==========================================================================================================
# VALIDATION PIPELINE
# ==========================================================================================================

class ValidationPipeline:
    """Multi-layer validation system"""

    def __init__(self):
        """Initialize with FIXED mango detector"""
        self.mango_detector = MangoDetector()

    def check_blur_adaptive(self, img: np.ndarray) -> Tuple[bool, float]:
        """Adaptive blur detection"""
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blur_score = cv2.Laplacian(gray, cv2.CV_64F).var()

        mean_brightness = gray.mean()
        threshold = Config.Validation.BLUR_THRESHOLD_BASE

        if mean_brightness < 100:
            threshold *= Config.Validation.BLUR_DARK_MULTIPLIER
        elif mean_brightness > 180:
            threshold *= Config.Validation.BLUR_BRIGHT_MULTIPLIER

        is_blurry = blur_score < threshold
        return is_blurry, blur_score

    def check_brightness(self, img: np.ndarray) -> Tuple[str, float]:
        """Validate lighting"""
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        brightness = hsv[:, :, 2].mean()

        if brightness < Config.Validation.BRIGHTNESS_MIN:
            return "too_dark", brightness
        elif brightness > Config.Validation.BRIGHTNESS_MAX:
            return "too_bright", brightness
        else:
            return "ok", brightness

    def check_duplicate(self, img1: np.ndarray, img2: np.ndarray) -> Tuple[bool, float]:
        """Check for duplicate images"""
        gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

        if gray1.shape != gray2.shape:
            gray2 = cv2.resize(gray2, (gray1.shape[1], gray1.shape[0]))

        ssim_score, _ = ssim(gray1, gray2, full=True)
        is_duplicate = ssim_score > Config.Validation.SSIM_DUPLICATE_THRESHOLD

        return is_duplicate, ssim_score

    def segment_mango(self, img: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
        """Segment mango from background"""
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        lower = np.array(Config.Features.HSV_LOWER)
        upper = np.array(Config.Features.HSV_UPPER)
        mask = cv2.inRange(hsv, lower, upper)
        kernel = np.ones((7, 7), np.uint8)
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
        mango = cv2.bitwise_and(img, img, mask=mask)
        return mango, mask

    def verify_same_mango(self, img1: np.ndarray, img2: np.ndarray) -> Tuple[bool, Dict]:
        """Triple-check same mango verification"""
        mango1, _ = self.segment_mango(img1)
        mango2, _ = self.segment_mango(img2)

        # Histogram correlation
        hsv1 = cv2.cvtColor(mango1, cv2.COLOR_BGR2HSV)
        hsv2 = cv2.cvtColor(mango2, cv2.COLOR_BGR2HSV)

        hist1 = cv2.calcHist([hsv1], [0, 1], None, [50, 60], [0, 180, 0, 256])
        hist2 = cv2.calcHist([hsv2], [0, 1], None, [50, 60], [0, 180, 0, 256])

        cv2.normalize(hist1, hist1)
        cv2.normalize(hist2, hist2)

        hist_corr = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)

        # SSIM
        g1 = cv2.cvtColor(mango1, cv2.COLOR_BGR2GRAY)
        g2 = cv2.cvtColor(mango2, cv2.COLOR_BGR2GRAY)
        g2 = cv2.resize(g2, (g1.shape[1], g1.shape[0]))
        ssim_score, _ = ssim(g1, g2, full=True)

        # ORB
        orb = cv2.ORB_create()
        kp1, des1 = orb.detectAndCompute(g1, None)
        kp2, des2 = orb.detectAndCompute(g2, None)

        matches = 0
        if des1 is not None and des2 is not None:
            bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
            matches = len(bf.match(des1, des2))

        # Decision
        check_hist = hist_corr >= Config.Validation.HISTOGRAM_CORRELATION_MIN
        check_ssim = ssim_score >= Config.Validation.SSIM_SAME_MANGO_MIN
        check_orb = matches >= Config.Validation.ORB_MATCHES_MIN

        is_same = check_hist and check_ssim and check_orb

        details = {
            'histogram_correlation': hist_corr,
            'ssim': ssim_score,
            'orb_matches': matches,
            'checks_passed': {
                'histogram': check_hist,
                'ssim': check_ssim,
                'orb': check_orb
            }
        }

        return is_same, details

    def validate_image_pair(self, front_path: str, back_path: str) -> Dict:
        """Complete validation pipeline"""

        result = {
            'passed': False,
            'checks': {},
            'errors': [],
            'images': None
        }

        # Load images
        img_front = cv2.imread(front_path)
        img_back = cv2.imread(back_path)

        if img_front is None or img_back is None:
            result['errors'].append("Failed to load images")
            return result

        # Resize
        img_front = cv2.resize(img_front, Config.Features.IMAGE_SIZE)
        img_back = cv2.resize(img_back, Config.Features.IMAGE_SIZE)

        # Mango detection
        mango_front = self.mango_detector.detect(img_front)
        mango_back = self.mango_detector.detect(img_back)

        result['checks']['mango_detection'] = {
            'front': mango_front,
            'back': mango_back
        }

        if not (mango_front['is_mango'] and mango_back['is_mango']):
            result['errors'].append(
                f"Mango detection failed "
                f"(Front: {mango_front['confidence']:.1%} via {mango_front['method']}, "
                f"Back: {mango_back['confidence']:.1%} via {mango_back['method']})"
            )
            return result

        # Blur check
        blur_front, score_front = self.check_blur_adaptive(img_front)
        blur_back, score_back = self.check_blur_adaptive(img_back)

        result['checks']['blur'] = {
            'front': {'is_blurry': blur_front, 'score': score_front},
            'back': {'is_blurry': blur_back, 'score': score_back}
        }

        if blur_front or blur_back:
            result['errors'].append(f"Blurry image (Front: {score_front:.1f}, Back: {score_back:.1f})")
            return result

        # Brightness check
        bright_front, val_front = self.check_brightness(img_front)
        bright_back, val_back = self.check_brightness(img_back)

        result['checks']['brightness'] = {
            'front': {'status': bright_front, 'value': val_front},
            'back': {'status': bright_back, 'value': val_back}
        }

        if bright_front != "ok" or bright_back != "ok":
            result['errors'].append(f"Lighting issue (Front: {bright_front}, Back: {bright_back})")
            return result

        # Duplicate check
        is_dup, dup_score = self.check_duplicate(img_front, img_back)

        result['checks']['duplicate'] = {
            'is_duplicate': is_dup,
            'ssim': dup_score
        }

        if is_dup:
            result['errors'].append(f"Duplicate images (SSIM: {dup_score:.3f})")
            return result

        # Same mango check
        is_same, same_details = self.verify_same_mango(img_front, img_back)

        result['checks']['same_mango'] = same_details

        if not is_same:
            failed = [k for k, v in same_details['checks_passed'].items() if not v]
            result['errors'].append(f"Different mangoes (Failed: {', '.join(failed)})")
            return result

        # All passed
        result['passed'] = True
        result['images'] = {'front': img_front, 'back': img_back}

        return result


# ==========================================================================================================
# FEATURE EXTRACTION
# ==========================================================================================================

def extract_standard_features(img: np.ndarray) -> np.ndarray:
    """Extract 37 standard features"""
    img = cv2.resize(img, Config.Features.IMAGE_SIZE)

    rgb_mean = img.mean(axis=(0, 1))
    rgb_std = img.std(axis=(0, 1))

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0, 1))
    hsv_std = hsv.std(axis=(0, 1))

    hue = hsv[:, :, 0]
    hue_hist, _ = np.histogram(hue, bins=16, range=(0, 180))
    hue_hist = hue_hist / (hue_hist.sum() + 1e-10)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_q = (gray / 4).astype(np.uint8)

    glcm = graycomatrix(gray_q, Config.Features.GLCM_DISTANCES, Config.Features.GLCM_ANGLES,
                        levels=Config.Features.GLCM_LEVELS, symmetric=True, normed=True)

    texture = [
        graycoprops(glcm, 'contrast').mean(),
        graycoprops(glcm, 'homogeneity').mean(),
        graycoprops(glcm, 'energy').mean(),
        graycoprops(glcm, 'correlation').mean(),
        graycoprops(glcm, 'dissimilarity').mean()
    ]

    return np.hstack([rgb_mean, rgb_std, hsv_mean, hsv_std, hue_hist, texture])


def extract_defect_features(img: np.ndarray) -> Dict:
    """Extract defect features"""
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    dark_spots = hsv[:, :, 2] < Config.Features.DARK_SPOT_THRESHOLD
    dark_density = dark_spots.sum() / dark_spots.size

    sat = hsv[:, :, 1]
    abnormal = ((sat < Config.Features.ABNORMAL_SAT_LOW) | (sat > Config.Features.ABNORMAL_SAT_HIGH))
    abnormal_ratio = abnormal.sum() / abnormal.size

    hue_std = hsv[:, :, 0].std()
    uniformity = 1.0 / (1.0 + hue_std / 50.0)

    return {
        'dark_spot_density': dark_density,
        'abnormal_pixel_ratio': abnormal_ratio,
        'color_uniformity': uniformity
    }


# ==========================================================================================================
# TRAINING
# ==========================================================================================================

print("\n" + "="*100)
print("ADVANCED AGENTIC AI + DECISION SUPPORT SYSTEM".center(100))
print("="*100)

print("\n" + "="*100)
print("[PIPELINE] LOADING & TRAINING GRADIENT BOOSTING MODELS".center(100))
print("="*100)

print("\n[1/5] Loading dataset...")
df = pd.read_excel(Config.EXCEL_PATH)
df["Fruit Grade"] = df["Fruit Grade"].astype(str)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")
print(f"✓ Loaded {len(df)} records")

print("\n[2/5] Extracting features...")
X_img, X_weight, y_rip, y_grd, groups = [], [], [], [], []

for fid, g in df.groupby("Fruit_ID"):
    if len(g) != 2:
        continue
    r1, r2 = g.iloc[0], g.iloc[1]
    img1 = cv2.imread(os.path.join(Config.IMAGE_DIR, r1["Fruit No"]))
    img2 = cv2.imread(os.path.join(Config.IMAGE_DIR, r2["Fruit No"]))
    if img1 is None or img2 is None:
        continue

    f1 = extract_standard_features(img1)
    f2 = extract_standard_features(img2)
    feat = (f1 + f2) / 2

    X_img.append(feat)
    X_weight.append([r1["Mass(kg)"]])
    y_rip.append(r1["Color_K-Yellow_P_Green"])
    y_grd.append(r1["Fruit Grade"])
    groups.append(fid)

X = np.hstack([np.array(X_img), np.array(X_weight)])
print(f"✓ Loaded {len(X)} samples")

print("\n[3/5] Encoding labels...")
rip_enc = LabelEncoder()
grd_enc = LabelEncoder()
y_rip = rip_enc.fit_transform(y_rip)
y_grd = grd_enc.fit_transform(y_grd)
print(f"✓ Ripeness: {rip_enc.classes_}")
print(f"✓ Grade: {grd_enc.classes_}")

print("\n[4/5] K-Fold Cross-Validation...")
kfold = StratifiedKFold(n_splits=Config.ML.CV_FOLDS, shuffle=True, random_state=Config.ML.CV_RANDOM_STATE)
cv_rip, cv_grd = [], []

for fold, (tr_idx, te_idx) in enumerate(kfold.split(X, y_rip), 1):
    X_tr, X_te = X[tr_idx], X[te_idx]
    yr_tr, yr_te = y_rip[tr_idx], y_rip[te_idx]
    yg_tr, yg_te = y_grd[tr_idx], y_grd[te_idx]

    scaler = RobustScaler()
    X_tr_s = scaler.fit_transform(X_tr)
    X_te_s = scaler.transform(X_te)

    gb_r = GradientBoostingClassifier(
        n_estimators=Config.ML.GB_N_ESTIMATORS, learning_rate=Config.ML.GB_LEARNING_RATE,
        max_depth=Config.ML.GB_MAX_DEPTH, min_samples_split=Config.ML.GB_MIN_SAMPLES_SPLIT,
        min_samples_leaf=Config.ML.GB_MIN_SAMPLES_LEAF, subsample=Config.ML.GB_SUBSAMPLE,
        random_state=Config.ML.GB_RANDOM_STATE
    )
    gb_g = GradientBoostingClassifier(
        n_estimators=Config.ML.GB_N_ESTIMATORS, learning_rate=Config.ML.GB_LEARNING_RATE,
        max_depth=Config.ML.GB_MAX_DEPTH, min_samples_split=Config.ML.GB_MIN_SAMPLES_SPLIT,
        min_samples_leaf=Config.ML.GB_MIN_SAMPLES_LEAF, subsample=Config.ML.GB_SUBSAMPLE,
        random_state=Config.ML.GB_RANDOM_STATE
    )

    gb_r.fit(X_tr_s, yr_tr)
    gb_g.fit(X_tr_s, yg_tr)

    cv_rip.append(f1_score(yr_te, gb_r.predict(X_te_s), average='weighted'))
    cv_grd.append(f1_score(yg_te, gb_g.predict(X_te_s), average='weighted'))

print(f"\n✓ Ripeness F1: {np.mean(cv_rip):.4f} ± {np.std(cv_rip):.4f}")
print(f"✓ Grade F1: {np.mean(cv_grd):.4f} ± {np.std(cv_grd):.4f}")

print("\n[5/5] Training final models...")
final_scaler = RobustScaler()
X_final = final_scaler.fit_transform(X)

final_gb_rip = GradientBoostingClassifier(
    n_estimators=Config.ML.GB_N_ESTIMATORS, learning_rate=Config.ML.GB_LEARNING_RATE,
    max_depth=Config.ML.GB_MAX_DEPTH, min_samples_split=Config.ML.GB_MIN_SAMPLES_SPLIT,
    min_samples_leaf=Config.ML.GB_MIN_SAMPLES_LEAF, subsample=Config.ML.GB_SUBSAMPLE,
    random_state=Config.ML.GB_RANDOM_STATE
)
final_gb_grd = GradientBoostingClassifier(
    n_estimators=Config.ML.GB_N_ESTIMATORS, learning_rate=Config.ML.GB_LEARNING_RATE,
    max_depth=Config.ML.GB_MAX_DEPTH, min_samples_split=Config.ML.GB_MIN_SAMPLES_SPLIT,
    min_samples_leaf=Config.ML.GB_MIN_SAMPLES_LEAF, subsample=Config.ML.GB_SUBSAMPLE,
    random_state=Config.ML.GB_RANDOM_STATE
)

final_gb_rip.fit(X_final, y_rip)
final_gb_grd.fit(X_final, y_grd)
print("✓ Models trained successfully")


# ==========================================================================================================
# DSS
# ==========================================================================================================

class DecisionSupportSystem:
    """DSS for shelf life and market routing - FIXED"""

    def __init__(self):
        pass

    def predict_shelf_life(self, rip_pred, defects, texture, rip_proba):
        """Predict shelf life with safe feature indexing"""

        rip_conf = rip_proba[rip_pred]
        rip_score = (1.0 - rip_pred) * rip_conf

        defect_score = 1.0 - (defects['dark_spot_density'] * 0.6 + defects['abnormal_pixel_ratio'] * 0.4)
        defect_score = np.clip(defect_score, 0, 1)

        texture_contrast = texture[32] if len(texture) > 32 else texture[-5]
        texture_dissimilarity = texture[36] if len(texture) > 36 else texture[-1]

        texture_score = np.clip((texture_contrast + texture_dissimilarity) / 2, 0, 1)

        final_score = (
            rip_score * Config.DSS.ShelfLife.WEIGHT_RIPENESS +
            defect_score * Config.DSS.ShelfLife.WEIGHT_DEFECTS +
            texture_score * Config.DSS.ShelfLife.WEIGHT_TEXTURE
        )

        days = Config.DSS.ShelfLife.MIN_DAYS + (Config.DSS.ShelfLife.MAX_DAYS - Config.DSS.ShelfLife.MIN_DAYS) * final_score

        return {
            'shelf_life_days': int(np.round(days)),
            'freshness_score': final_score,
            'component_scores': {'ripeness': rip_score, 'defects': defect_score, 'texture': texture_score}
        }

    def recommend_market_routing(self, grd_pred, shelf_days, weight, grd_proba):
        """Market routing recommendation"""

        grade_map = {0: 1.0, 1: 0.65, 2: 0.3}
        grd_score = grade_map.get(grd_pred, 0.3) * grd_proba[grd_pred]

        shelf_score = np.clip(shelf_days / Config.DSS.ShelfLife.MAX_DAYS, 0, 1)

        if Config.DSS.MarketRouting.OPTIMAL_WEIGHT_MIN <= weight <= Config.DSS.MarketRouting.OPTIMAL_WEIGHT_MAX:
            weight_score = 1.0
        elif weight < Config.DSS.MarketRouting.OPTIMAL_WEIGHT_MIN:
            weight_score = 0.7
        else:
            weight_score = 0.85

        composite = (
            grd_score * Config.DSS.MarketRouting.WEIGHT_GRADE +
            shelf_score * Config.DSS.MarketRouting.WEIGHT_SHELF_LIFE +
            weight_score * Config.DSS.MarketRouting.WEIGHT_WEIGHT
        )

        if composite >= Config.DSS.MarketRouting.EXPORT_THRESHOLD:
            rec = "EXPORT"
            reason = f"Premium quality (Score: {composite:.2f})"
        elif composite >= Config.DSS.MarketRouting.RETAIL_THRESHOLD:
            rec = "RETAIL"
            reason = f"Good retail quality (Score: {composite:.2f})"
        else:
            rec = "PROCESSING"
            reason = f"Processing suitable (Score: {composite:.2f})"

        return {
            'recommendation': rec,
            'composite_score': composite,
            'component_scores': {'grade': grd_score, 'shelf_life': shelf_score, 'weight': weight_score},
            'reasoning': reason
        }

    def generate_explanation(self, rip_pred, grd_pred, rip_proba, grd_proba, defects, shelf_result, rip_classes, grd_classes):
        """Generate confidence and reasoning"""

        rip_conf = rip_proba[rip_pred]
        grd_conf = grd_proba[grd_pred]
        overall = (rip_conf + grd_conf) / 2

        reasoning = []
        if rip_conf > 0.8:
            reasoning.append(f"High ripeness confidence: {rip_classes[rip_pred]} ({rip_conf:.1%})")
        if defects['dark_spot_density'] > 0.1:
            reasoning.append(f"Dark spots detected ({defects['dark_spot_density']:.1%})")
        if shelf_result['freshness_score'] > 0.7:
            reasoning.append("High freshness indicates extended shelf life")

        return {
            'overall_confidence': overall,
            'ripeness_confidence': rip_conf,
            'grade_confidence': grd_conf,
            'reasoning': "; ".join(reasoning) if reasoning else "Standard assessment",
            'prediction_probabilities': {
                'ripeness': {str(rip_classes[i]): float(p) for i, p in enumerate(rip_proba)},
                'grade': {str(grd_classes[i]): float(p) for i, p in enumerate(grd_proba)}
            }
        }
# ==========================================================================================================
# PREDICTION FUNCTION
# ==========================================================================================================

def predict_mango_advanced(front_path: str, back_path: str, weight_grams: Optional[float] = None) -> Dict:
    """Complete prediction pipeline"""

    print("\n" + "="*100)
    print("🍋 ADVANCED MANGO QUALITY ASSESSMENT".center(100))
    print("="*100)

    if weight_grams is None:
        weight_grams = float(input("\n📝 Enter mango weight (grams): "))

    validator = ValidationPipeline()
    dss = DecisionSupportSystem()

    print("\n" + "-"*100)
    print("STAGE 1: MULTI-LAYER VALIDATION".center(100))
    print("-"*100)

    validation = validator.validate_image_pair(front_path, back_path)

    if not validation['passed']:
        print("\n❌ VALIDATION FAILED:")
        for err in validation['errors']:
            print(f"   - {err}")
        return {'status': 'FAILED', 'validation': validation}

    print("✅ All validation checks passed")

    print("\n" + "-"*100)
    print("STAGE 2: FEATURE EXTRACTION".center(100))
    print("-"*100)

    img_f = validation['images']['front']
    img_b = validation['images']['back']

    feat = (extract_standard_features(img_f) + extract_standard_features(img_b)) / 2
    defects = {
        k: (extract_defect_features(img_f)[k] + extract_defect_features(img_b)[k]) / 2
        for k in ['dark_spot_density', 'abnormal_pixel_ratio', 'color_uniformity']
    }

    print("✓ Extracted features")

    print("\n" + "-"*100)
    print("STAGE 3: GRADIENT BOOSTING PREDICTION".center(100))
    print("-"*100)

    X_in = np.hstack([feat, [weight_grams]]).reshape(1, -1)
    X_in_s = final_scaler.transform(X_in)

    rip_pred = final_gb_rip.predict(X_in_s)[0]
    rip_proba = final_gb_rip.predict_proba(X_in_s)[0]
    rip_class = rip_enc.inverse_transform([rip_pred])[0]

    grd_pred = final_gb_grd.predict(X_in_s)[0]
    grd_proba = final_gb_grd.predict_proba(X_in_s)[0]
    grd_class = grd_enc.inverse_transform([grd_pred])[0]

    print(f"✓ Ripeness: {rip_class} ({rip_proba[rip_pred]:.1%})")
    print(f"✓ Grade: {grd_class} ({grd_proba[grd_pred]:.1%})")

    print("\n" + "-"*100)
    print("STAGE 4: DECISION SUPPORT ANALYSIS".center(100))
    print("-"*100)

    shelf = dss.predict_shelf_life(rip_pred, defects, feat, rip_proba)
    routing = dss.recommend_market_routing(grd_pred, shelf['shelf_life_days'], weight_grams, grd_proba)
    explanation = dss.generate_explanation(rip_pred, grd_pred, rip_proba, grd_proba, defects, shelf, rip_enc.classes_, grd_enc.classes_)

    print(f"✓ Shelf Life: {shelf['shelf_life_days']} days")
    print(f"✓ Market: {routing['recommendation']}")

    print("\n" + "="*100)
    print("COMPREHENSIVE ASSESSMENT REPORT".center(100))
    print("="*100)

    print(f"\n{'QUALITY ASSESSMENT':-^100}")
    print(f"  Ripeness: {rip_class} ({explanation['ripeness_confidence']:.1%})")
    print(f"  Grade: {grd_class} ({explanation['grade_confidence']:.1%})")
    print(f"  Weight: {weight_grams}g")

    print(f"\n{'DECISION SUPPORT':-^100}")
    print(f"  Shelf Life: {shelf['shelf_life_days']} days")
    print(f"  Market: {routing['recommendation']}")
    print(f"  Score: {routing['composite_score']:.2f}")

    print(f"\n{'DEFECT ANALYSIS':-^100}")
    print(f"  Dark Spots: {defects['dark_spot_density']:.2%}")
    print(f"  Abnormal Pixels: {defects['abnormal_pixel_ratio']:.2%}")
    print(f"  Uniformity: {defects['color_uniformity']:.2f}")

    print(f"\n{'REASONING':-^100}")
    print(f"  {explanation['reasoning']}")
    print("\n" + "="*100)

    return {
        'status': 'SUCCESS',
        'quality_assessment': {'ripeness': rip_class, 'grade': grd_class, 'weight_grams': weight_grams},
        'confidence_scores': explanation,
        'decision_support': {'shelf_life_days': shelf['shelf_life_days'], 'market_recommendation': routing['recommendation']},
        'defect_analysis': defects
    }
```

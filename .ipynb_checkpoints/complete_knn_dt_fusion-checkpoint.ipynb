```

import numpy as np
import pandas as pd

from sklearn.model_selection import GroupShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```

# ===============================
# REPLACE THIS SECTION WITH YOUR REAL DATA
# ===============================

np.random.seed(42)
X = np.random.rand(200, 21)   # 20 image features + 1 weight
y_r = np.random.randint(0,2,200)   # Ripeness
y_g = np.random.randint(0,3,200)   # Grade
groups = np.random.randint(0,50,200)
```

```

gss = GroupShuffleSplit(test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y_r, groups))

X_train, X_test = X[train_idx], X[test_idx]
yr_train, yr_test = y_r[train_idx], y_r[test_idx]
yg_train, yg_test = y_g[train_idx], y_g[test_idx]

X_img = X[:, :-1]
X_w = X[:, -1:].reshape(-1,1)

X_img_train, X_img_test = X_img[train_idx], X_img[test_idx]
X_w_train, X_w_test = X_w[train_idx], X_w[test_idx]
```

```

def compute_metrics(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_true, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(y_true, y_pred, average="weighted", zero_division=0)
    }
```

```

results = []

# EARLY FUSION
knn = KNeighborsClassifier(5)
dt = DecisionTreeClassifier(max_depth=5, random_state=42)

# KNN Early
knn.fit(X_train, yr_train)
results.append({"Model":"KNN","Fusion":"Early","Task":"Ripeness", **compute_metrics(yr_test, knn.predict(X_test))})
knn.fit(X_train, yg_train)
results.append({"Model":"KNN","Fusion":"Early","Task":"Grade", **compute_metrics(yg_test, knn.predict(X_test))})

# DT Early
dt.fit(X_train, yr_train)
results.append({"Model":"Decision Tree","Fusion":"Early","Task":"Ripeness", **compute_metrics(yr_test, dt.predict(X_test))})
dt.fit(X_train, yg_train)
results.append({"Model":"Decision Tree","Fusion":"Early","Task":"Grade", **compute_metrics(yg_test, dt.predict(X_test))})

# LATE FUSION
def late(model, X_img_tr, X_w_tr, X_img_te, X_w_te, y_tr):
    m1 = model()
    m2 = model()
    m1.fit(X_img_tr, y_tr)
    m2.fit(X_w_tr, y_tr)
    prob = (m1.predict_proba(X_img_te) + m2.predict_proba(X_w_te))/2
    return np.argmax(prob, axis=1)

# KNN Late
results.append({"Model":"KNN","Fusion":"Late","Task":"Ripeness",
                **compute_metrics(yr_test, late(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"KNN","Fusion":"Late","Task":"Grade",
                **compute_metrics(yg_test, late(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# DT Late
results.append({"Model":"Decision Tree","Fusion":"Late","Task":"Ripeness",
                **compute_metrics(yr_test, late(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Decision Tree","Fusion":"Late","Task":"Grade",
                **compute_metrics(yg_test, late(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# HYBRID FUSION
def hybrid(model, X_img_tr, X_w_tr, X_img_te, X_w_te, y_tr):
    m1 = model()
    m2 = model()
    m1.fit(X_img_tr, y_tr)
    m2.fit(X_w_tr, y_tr)
    prob = 0.6*m1.predict_proba(X_img_te) + 0.4*m2.predict_proba(X_w_te)
    return np.argmax(prob, axis=1)

# KNN Hybrid
results.append({"Model":"KNN","Fusion":"Hybrid","Task":"Ripeness",
                **compute_metrics(yr_test, hybrid(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"KNN","Fusion":"Hybrid","Task":"Grade",
                **compute_metrics(yg_test, hybrid(lambda:KNeighborsClassifier(5),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})

# DT Hybrid
results.append({"Model":"Decision Tree","Fusion":"Hybrid","Task":"Ripeness",
                **compute_metrics(yr_test, hybrid(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yr_train))})
results.append({"Model":"Decision Tree","Fusion":"Hybrid","Task":"Grade",
                **compute_metrics(yg_test, hybrid(lambda:DecisionTreeClassifier(max_depth=5,random_state=42),
                X_img_train,X_w_train,X_img_test,X_w_test,yg_train))})
```

```

results_df = pd.DataFrame(results).round(4)
print(results_df)
results_df.to_excel("final_fusion_results.xlsx", index=False)
print("Saved as final_fusion_results.xlsx")
```

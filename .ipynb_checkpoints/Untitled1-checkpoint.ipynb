```python
# ==========================================
# MANGO MATURITY — FUSION MODEL COMPARISON
# USING GROUP SHUFFLE SPLIT (NO DATA LEAKAGE)
# ==========================================

import os
import numpy as np
import pandas as pd
from PIL import Image
from scipy.stats import mode

from sklearn.model_selection import GroupShuffleSplit
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler

# ======================
# FILE PATHS
# ======================
EXCEL_PATH = "/content/Maturity_dataset.xlsx"
IMAGE_FOLDER = "/content/images"

# ======================
# LOAD DATA
# ======================
df = pd.read_excel(EXCEL_PATH)
has_weight = "Weight" in df.columns

def load_image(path, size=(64,64)):
    img = Image.open(path).convert("RGB")
    img = img.resize(size)
    return np.array(img).flatten()

# ======================
# FEATURE EXTRACTION
# ======================
img_features, weights, labels, groups = [], [], [], []

for _, row in df.iterrows():
    file_name = row["Fruit No"]
    path = os.path.join(IMAGE_FOLDER, file_name)

    if os.path.exists(path):
        img_features.append(load_image(path))
        labels.append(row["Maturity"])
        if has_weight:
            weights.append(row["Weight"])

        # extract fruit group (e.g., 1a.jpg → 1)
        group_id = ''.join(filter(str.isdigit, file_name))
        groups.append(group_id)

if not img_features:
    print("❌ Error: No images found. Check IMAGE_FOLDER.")
elif not has_weight:
    print("⚠️ Warning: 'Weight' column not found. Performing Image-only classification.")
    X_img = np.array(img_features)
    y = np.array(labels)
    groups = np.array(groups)
    gss = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
    train_idx, test_idx = next(gss.split(X_img, y, groups))
    Xi_train, Xi_test = X_img[train_idx], X_img[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    model = RandomForestClassifier(n_estimators=100, random_state=42).fit(Xi_train, y_train)
    acc = accuracy_score(y_test, model.predict(Xi_test))
    print(f"Image-only Accuracy: {acc:.2f}")
else:
    X_img = np.array(img_features)
    X_weight = StandardScaler().fit_transform(np.array(weights).reshape(-1,1))
    y = np.array(labels)
    groups = np.array(groups)

    gss = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
    train_idx, test_idx = next(gss.split(X_img, y, groups))

    Xi_train, Xi_test = X_img[train_idx], X_img[test_idx]
    Xw_train, Xw_test = X_weight[train_idx], X_weight[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    Xe_train, Xe_test = np.hstack((Xi_train, Xw_train)), np.hstack((Xi_test, Xw_test))

    def evaluate(y_true, pred):
        return (
            accuracy_score(y_true, pred),
            precision_score(y_true, pred, average="weighted", zero_division=0),
            recall_score(y_true, pred, average="weighted", zero_division=0),
            f1_score(y_true, pred, average="weighted", zero_division=0),
        )

    results = []
    # Early Fusion
    for name, model in {"Naive Bayes": GaussianNB(), "Random Forest": RandomForestClassifier(random_state=42)}.items():
        model.fit(Xe_train, y_train)
        results.append([name, "Early Fusion", "Maturity", *evaluate(y_test, model.predict(Xe_test))])

    # Late Fusion
    m_img = RandomForestClassifier(random_state=42).fit(Xi_train, y_train)
    m_w = GaussianNB().fit(Xw_train, y_train)
    late_pred = mode(np.vstack((m_img.predict(Xi_test), m_w.predict(Xw_test))), axis=0)[0].flatten()
    results.append(["Fusion Model", "Late Fusion", "Maturity", *evaluate(y_test, late_pred)])

    print(pd.DataFrame(results, columns=["Model", "Fusion", "Task", "Accuracy", "Precision", "Recall", "F1"]))
```

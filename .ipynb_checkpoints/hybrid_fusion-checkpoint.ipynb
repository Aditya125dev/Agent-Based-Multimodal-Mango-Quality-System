# Hybrid Fusion Logistic Regression Notebook

```python
import os
import cv2
import numpy as np
import pandas as pd
from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

IMAGE_DIR='images'
EXCEL_PATH='Harumanis_mango_weight_grade.xlsx'
df=pd.read_excel(EXCEL_PATH)
df['Fruit_ID']=df['Fruit No'].str.extract(r'(\d+)')
```

```python
def extract_image_features(img):
    img=cv2.resize(img,(224,224))
    rgb_mean=img.mean(axis=(0,1))
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    glcm=graycomatrix((gray/4).astype('uint8'),[1],[0],levels=64,symmetric=True,normed=True)
    contrast=graycoprops(glcm,'contrast')[0,0]
    return np.hstack([rgb_mean,contrast])
```

```python
X_img=[]; X_weight=[]; y=[]; groups=[]
for _,row in df.iterrows():
    path=os.path.join(IMAGE_DIR,row['Fruit No'])
    img=cv2.imread(path)
    if img is None: continue
    X_img.append(extract_image_features(img))
    X_weight.append([row['Mass(kg)']])
    y.append(row['Fruit Grade'])
    groups.append(row['Fruit_ID'])
X_img=np.array(X_img); X_weight=np.array(X_weight)
X_comb=np.hstack([X_img,X_weight])
le=LabelEncoder(); y=le.fit_transform(y); groups=np.array(groups)
```

```python
gss=GroupShuffleSplit(test_size=0.2,random_state=42)
train_idx,test_idx=next(gss.split(X_img,y,groups))
Xi_tr,Xi_te=X_img[train_idx],X_img[test_idx]
Xw_tr,Xw_te=X_weight[train_idx],X_weight[test_idx]
Xc_tr,Xc_te=X_comb[train_idx],X_comb[test_idx]
y_tr,y_te=y[train_idx],y[test_idx]
```

```python
sc_i=RobustScaler(); sc_w=RobustScaler(); sc_c=RobustScaler()
Xi_tr=sc_i.fit_transform(Xi_tr); Xi_te=sc_i.transform(Xi_te)
Xw_tr=sc_w.fit_transform(Xw_tr); Xw_te=sc_w.transform(Xw_te)
Xc_tr=sc_c.fit_transform(Xc_tr); Xc_te=sc_c.transform(Xc_te)
```

```python
lr_i=LogisticRegression(max_iter=2000)
lr_w=LogisticRegression(max_iter=2000)
lr_c=LogisticRegression(max_iter=2000)
lr_i.fit(Xi_tr,y_tr); lr_w.fit(Xw_tr,y_tr); lr_c.fit(Xc_tr,y_tr)
```

```python
p_i=lr_i.predict_proba(Xi_te)
p_w=lr_w.predict_proba(Xw_te)
p_c=lr_c.predict_proba(Xc_te)
final_prob=0.4*p_i+0.3*p_w+0.3*p_c
y_pred=np.argmax(final_prob,axis=1)
print('Accuracy:',accuracy_score(y_te,y_pred))
print('Precision:',precision_score(y_te,y_pred,average='weighted'))
print('Recall:',recall_score(y_te,y_pred,average='weighted'))
print('F1:',f1_score(y_te,y_pred,average='weighted'))
```

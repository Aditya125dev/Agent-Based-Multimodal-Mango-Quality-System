```python

import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix


IMAGE_DIR = "/content/images"
EXCEL_PATH = "/content/Harumanis_mango_weight_grade.xlsx"

df = pd.read_excel(EXCEL_PATH)
df["Fruit Grade"] = df["Fruit Grade"].astype(str)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\\d+)")

# Check distributions
print("=== ORIGINAL DATA DISTRIBUTION ===")
print("\\nRipeness distribution:")
print(df['Color_K-Yellow_P_Green'].value_counts())
print("\\nGrade distribution:")
print(df['Fruit Grade'].value_counts())

def extract_image_features(img):
    img = cv2.resize(img, (224, 224))
    
    rgb_mean = img.mean(axis=(0,1))
    rgb_std = img.std(axis=(0,1))
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0,1))
    hsv_std = hsv.std(axis=(0,1))
    
    hue_channel = hsv[:, :, 0]
    hue_hist, _ = np.histogram(hue_channel, bins=16, range=(0, 180))
    hue_hist = hue_hist / hue_hist.sum()
    
    yellow_mask = (hue_channel >= 20) & (hue_channel <= 35)
    green_mask = (hue_channel >= 35) & (hue_channel <= 85)
    yellow_ratio = yellow_mask.sum() / (hue_channel.shape[0] * hue_channel.shape[1])
    green_ratio = green_mask.sum() / (hue_channel.shape[0] * hue_channel.shape[1])
    
    saturation_mean = hsv[:, :, 1].mean()
    saturation_std = hsv[:, :, 1].std()
    value_mean = hsv[:, :, 2].mean()
    value_std = hsv[:, :, 2].std()
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_reduced = (gray / 4).astype(np.uint8)
    
    glcm = graycomatrix(gray_reduced, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], 
                        levels=64, symmetric=True, normed=True)
    
    contrast = graycoprops(glcm, 'contrast').mean()
    homogeneity = graycoprops(glcm, 'homogeneity').mean()
    energy = graycoprops(glcm, 'energy').mean()
    correlation = graycoprops(glcm, 'correlation').mean()
    dissimilarity = graycoprops(glcm, 'dissimilarity').mean()
    
    edges = cv2.Canny(gray, 50, 150)
    edge_density = edges.sum() / (edges.shape[0] * edges.shape[1])
    
    features = np.hstack([
        rgb_mean, rgb_std,
        hsv_mean, hsv_std,
        [yellow_ratio, green_ratio],
        [saturation_mean, saturation_std, value_mean, value_std],
        hue_hist,
        [contrast, homogeneity, energy, correlation, dissimilarity, edge_density]
    ])
    
    return features


X, y_ripeness, y_grade, groups, meta = [], [], [], [], []

for fruit_id, g in df.groupby("Fruit_ID"):
    if len(g) != 2:
        continue
    
    row_a, row_b = g.iloc[0], g.iloc[1]
    
    img_a = cv2.imread(os.path.join(IMAGE_DIR, row_a["Fruit No"]))
    img_b = cv2.imread(os.path.join(IMAGE_DIR, row_b["Fruit No"]))
    
    if img_a is None or img_b is None:
        print(f"Warning: Could not load images for fruit {fruit_id}")
        continue
    
    f_a = extract_image_features(img_a)
    f_b = extract_image_features(img_b)
    
    f_mean = (f_a + f_b) / 2
    weight = np.array([row_a["Mass(kg)"]])
    features = np.hstack([f_mean, weight])
    
    X.append(features)
    y_ripeness.append(row_a["Color_K-Yellow_P_Green"])
    y_grade.append(row_a["Fruit Grade"])
    groups.append(fruit_id)
    
    meta.append({
        "fruit_id": fruit_id,
        "img_a": row_a["Fruit No"],
        "img_b": row_b["Fruit No"],
        "ripeness": row_a["Color_K-Yellow_P_Green"],
        "grade": row_a["Fruit Grade"],
        "weight": row_a["Mass(kg)"]
    })

X = np.array(X)
y_ripeness_str = np.array(y_ripeness)
y_grade_str = np.array(y_grade)
groups = np.array(groups)
meta = pd.DataFrame(meta)

print(f"\\n=== LOADED DATA ===")
print(f"Total samples: {len(X)}")
print(f"Feature dimension: {X.shape[1]}")

ripeness_enc = LabelEncoder()
grade_enc = LabelEncoder()
y_ripeness = ripeness_enc.fit_transform(y_ripeness_str)
y_grade = grade_enc.fit_transform(y_grade_str)

gss = GroupShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(X, y_ripeness, groups))

X_train, X_test = X[train_idx], X[test_idx]
yr_train, yr_test = y_ripeness[train_idx], y_ripeness[test_idx]
yg_train, yg_test = y_grade[train_idx], y_grade[test_idx]

scaler = RobustScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

lr_ripeness = LogisticRegression(C=0.1, max_iter=2000, class_weight='balanced')
lr_grade = LogisticRegression(C=0.1, max_iter=2000, multi_class="multinomial", class_weight='balanced')

lr_ripeness.fit(X_train_s, yr_train)
lr_grade.fit(X_train_s, yg_train)

yr_pred = lr_ripeness.predict(X_test_s)
yg_pred = lr_grade.predict(X_test_s)

print("\\nRIPENESS RESULTS")
print(classification_report(yr_test, yr_pred, zero_division=0))

print("\\nGRADE RESULTS")
print(classification_report(yg_test, yg_pred, zero_division=0))
```

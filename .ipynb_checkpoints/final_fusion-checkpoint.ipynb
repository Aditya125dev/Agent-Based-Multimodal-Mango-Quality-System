```

# ======================================================
# CONTINUATION: ADVANCED FUSION (KNN + DECISION TREE)
# ======================================================

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_decomposition import CCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import numpy as np
```

```

results = []

def store(model, fusion, task, y_true, y_pred):
    results.append({
        "Model": model,
        "Fusion": fusion,
        "Task": task,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(y_true, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(y_true, y_pred, average="weighted", zero_division=0)
    })
```

```

# ======================
# EARLY FUSION
# ======================

X_train_early = np.hstack([X_img_train, X_w_train])
X_test_early = np.hstack([X_img_test, X_w_test])

for name, model in [
    ("KNN", KNeighborsClassifier(5)),
    ("Decision Tree", DecisionTreeClassifier(max_depth=5, random_state=42))
]:
    model.fit(X_train_early, yr_train)
    store(name, "Early", "Ripeness", yr_test, model.predict(X_test_early))

    model.fit(X_train_early, yg_train)
    store(name, "Early", "Grade", yg_test, model.predict(X_test_early))
```

```

# ======================
# ENSEMBLE LEVEL
# ======================

def ensemble(model_class, y_train):
    m1 = model_class()
    m2 = model_class()

    m1.fit(X_img_train, y_train)
    m2.fit(X_w_train, y_train)

    p1 = m1.predict_proba(X_img_test)
    p2 = m2.predict_proba(X_w_test)

    return np.argmax((p1 + p2) / 2, axis=1)

store("KNN", "Ensemble Level", "Ripeness",
      yr_test, ensemble(lambda: KNeighborsClassifier(5), yr_train))

store("KNN", "Ensemble Level", "Grade",
      yg_test, ensemble(lambda: KNeighborsClassifier(5), yg_train))

store("Decision Tree", "Ensemble Level", "Ripeness",
      yr_test, ensemble(lambda: DecisionTreeClassifier(max_depth=5, random_state=42), yr_train))

store("Decision Tree", "Ensemble Level", "Grade",
      yg_test, ensemble(lambda: DecisionTreeClassifier(max_depth=5, random_state=42), yg_train))
```

```

# ======================
# CCA FUSION
# ======================

cca = CCA(n_components=2)

Xi_tr, Xw_tr = cca.fit_transform(X_img_train, X_w_train)
Xi_te, Xw_te = cca.transform(X_img_test, X_w_test)

X_train_cca = np.hstack([Xi_tr, Xw_tr])
X_test_cca = np.hstack([Xi_te, Xw_te])

for name, model in [
    ("KNN", KNeighborsClassifier(5)),
    ("Decision Tree", DecisionTreeClassifier(max_depth=5, random_state=42))
]:
    model.fit(X_train_cca, yr_train)
    store(name, "CCA Fusion", "Ripeness", yr_test, model.predict(X_test_cca))

    model.fit(X_train_cca, yg_train)
    store(name, "CCA Fusion", "Grade", yg_test, model.predict(X_test_cca))
```

```

# ======================
# STACKING (META-LEVEL)
# ======================

def stacking(model_class, y_train):
    m1 = model_class()
    m2 = model_class()

    m1.fit(X_img_train, y_train)
    m2.fit(X_w_train, y_train)

    meta_train = np.hstack([
        m1.predict_proba(X_img_train),
        m2.predict_proba(X_w_train)
    ])

    meta_model = LogisticRegression(max_iter=1000)
    meta_model.fit(meta_train, y_train)

    meta_test = np.hstack([
        m1.predict_proba(X_img_test),
        m2.predict_proba(X_w_test)
    ])

    return meta_model.predict(meta_test)

store("KNN", "Stacking", "Ripeness",
      yr_test, stacking(lambda: KNeighborsClassifier(5), yr_train))

store("KNN", "Stacking", "Grade",
      yg_test, stacking(lambda: KNeighborsClassifier(5), yg_train))

store("Decision Tree", "Stacking", "Ripeness",
      yr_test, stacking(lambda: DecisionTreeClassifier(max_depth=5, random_state=42), yr_train))

store("Decision Tree", "Stacking", "Grade",
      yg_test, stacking(lambda: DecisionTreeClassifier(max_depth=5, random_state=42), yg_train))
```

```

# ======================
# FINAL RESULTS TABLE
# ======================

results_df = pd.DataFrame(results).round(4)
results_df.to_excel("final_fusion_results.xlsx", index=False)
results_df
```

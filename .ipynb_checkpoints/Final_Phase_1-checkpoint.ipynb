# SVM And Logistic Regression

```python
# FRUIT MATURITY CLASSIFICATION — AUGMENTATION ONLY

import os
import cv2
import numpy as np
import pandas as pd

from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier


# PATHS

IMAGE_DIR = "/content/images"
LABEL_FILE = "/content/Maturity_dataset.xlsx"


# LOAD & ALIGN DATASET

df = pd.read_excel(LABEL_FILE)

df = df[df["Fruit No"].apply(
    lambda x: os.path.exists(os.path.join(IMAGE_DIR, x))
)].reset_index(drop=True)

df["Label"] = df["Maturity"].map({"Unripe": 0, "Ripe": 1})
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)").astype(int)

# LOAD IMAGES (RESIZED)

images, labels = [], []

for _, row in df.iterrows():
    img = cv2.imread(os.path.join(IMAGE_DIR, row["Fruit No"]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))
    images.append(img)
    labels.append(row["Label"])

images = np.array(images)
labels = np.array(labels)

# FRUIT-LEVEL SPLIT

gss = GroupShuffleSplit(test_size=0.4, random_state=42)
train_idx, test_idx = next(gss.split(images, labels, groups=df["Fruit_ID"]))

X_train_imgs = images[train_idx]
y_train = labels[train_idx]

X_test_imgs = images[test_idx]
y_test = labels[test_idx]

print("Train distribution:", np.unique(y_train, return_counts=True))
print("Test distribution :", np.unique(y_test, return_counts=True))

# FEATURE EXTRACTION (COLOR + TEXTURE)

def extract_features(img):
    rgb_mean = np.mean(img, axis=(0, 1))
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    hsv_mean = np.mean(hsv, axis=(0, 1))

    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gray = (gray / 4).astype(np.uint8)  # 64 gray levels

    glcm = graycomatrix(
        gray, [1], [0],
        levels=64,
        symmetric=True,
        normed=True
    )

    contrast = graycoprops(glcm, 'contrast')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]

    return np.hstack([
        rgb_mean,
        hsv_mean,
        contrast,
        homogeneity,
        energy,
        correlation
    ])

X_train = np.array([extract_features(img) for img in X_train_imgs])
X_test  = np.array([extract_features(img) for img in X_test_imgs])


# BASELINE RANDOM FOREST (NO AUGMENTATION)

rf_base = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    min_samples_leaf=5,
    random_state=42
)

rf_base.fit(X_train, y_train)
y_pred_base = rf_base.predict(X_test)

acc_base = accuracy_score(y_test, y_pred_base)
prec_base = precision_score(y_test, y_pred_base)
rec_base = recall_score(y_test, y_pred_base)
f1_base = f1_score(y_test, y_pred_base)

results.append({
    'Phase': 'Phase 1',
    'Model': 'Random Forest',
    'Task': 'Maturity Classification',
    'Fusion': 'Image Only',
    'Type': 'Baseline',
    'Parameters': str(rf_base.get_params()),
    'Accuracy': acc_base,
    'Precision': prec_base,
    'Recall': rec_base,
    'F1': f1_base
})

print("\nBASELINE RANDOM FOREST")
print("Accuracy :", acc_base)
print("Precision:", prec_base)
print("Recall   :", rec_base)
print("F1       :", f1_base)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_base))


# AUGMENTATION (TRAIN SET ONLY)

def augment_image(img):
    img = img.astype(np.float32)

    # Brightness
    img *= np.random.uniform(0.7, 1.3)

    # Small rotation
    angle = np.random.uniform(-15, 15)
    h, w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w//2, h//2), angle, 1)
    img = cv2.warpAffine(img, M, (w, h), borderMode=cv2.BORDER_REFLECT)

    # Slight blur
    if np.random.rand() < 0.5:
        img = cv2.GaussianBlur(img, (3,3), 0)

    return np.clip(img, 0, 255).astype(np.uint8)

X_train_aug, y_train_aug = [], []

for img, label in zip(X_train_imgs, y_train):
    # Original
    X_train_aug.append(extract_features(img))
    y_train_aug.append(label)

    # Augmented
    aug_img = augment_image(img)
    X_train_aug.append(extract_features(aug_img))
    y_train_aug.append(label)

X_train_aug = np.array(X_train_aug)
y_train_aug = np.array(y_train_aug)

print("\nAugmented training size:", X_train_aug.shape)


# RANDOM FOREST TRAINED ON AUGMENTED DATA

rf_aug = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    min_samples_leaf=5,
    random_state=42
)

rf_aug.fit(X_train_aug, y_train_aug)
y_pred_aug = rf_aug.predict(X_test)

acc_aug = accuracy_score(y_test, y_pred_aug)
prec_aug = precision_score(y_test, y_pred_aug)
rec_aug = recall_score(y_test, y_pred_aug)
f1_aug = f1_score(y_test, y_pred_aug)

results.append({
    'Phase': 'Phase 1',
    'Model': 'Random Forest',
    'Task': 'Maturity Classification',
    'Fusion': 'Image Only',
    'Type': 'Augmented',
    'Parameters': str(rf_aug.get_params()),
    'Accuracy': acc_aug,
    'Precision': prec_aug,
    'Recall': rec_aug,
    'F1': f1_aug
})

print("\nAUGMENTED RANDOM FOREST")
print("Accuracy :", acc_aug)
print("Precision:", prec_aug)
print("Recall   :", rec_aug)
print("F1       :", f1_aug)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_aug))
```

# Gradient Boosting And XGBoost


Fruit maturity and grade classififcation


```python
import os
import cv2
import warnings
import numpy as np
import pandas as pd
import time

from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import LabelEncoder

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input

warnings.filterwarnings("ignore")

# PATHS
IMAGE_DIR = "/content/images"
EXCEL_PATH = "/content/Maturity_dataset.xlsx"

# LOAD DATA
df = pd.read_excel(EXCEL_PATH)
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

# LOAD IMAGES
images, labels, group_ids = [], [], []

for _, row in df.iterrows():
    img_path = os.path.join(IMAGE_DIR, row["Fruit No"])
    img = cv2.imread(img_path)

    if img is None:
        continue

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))

    images.append(img)
    labels.append(row["Maturity"])
    group_ids.append(row["Fruit_ID"])

images = np.array(images, dtype=np.float32)
labels = np.array(labels)
group_ids = np.array(group_ids)

# FEATURE EXTRACTION (ResNet50)
resnet = ResNet50(weights="imagenet", include_top=False, pooling="avg")
images_pp = preprocess_input(images * 255)
features = resnet.predict(images_pp, verbose=0)

# ENCODE LABELS
le = LabelEncoder()
y = le.fit_transform(labels)

# GROUP SPLIT
gss = GroupShuffleSplit(test_size=0.3, random_state=42)
train_idx, test_idx = next(gss.split(features, y, groups=group_ids))

X_train, X_test = features[train_idx], features[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

# PARAMETER SETS
gb_params = [
    {'n_estimators': 50, 'learning_rate': 0.1, 'max_depth': 3},
    {'n_estimators': 100, 'learning_rate': 0.1, 'max_depth': 5},
    {'n_estimators': 150, 'learning_rate': 0.05, 'max_depth': 5},
]

xgb_params = [
    {'n_estimators': 50, 'learning_rate': 0.1, 'max_depth': 3},
    {'n_estimators': 100, 'learning_rate': 0.1, 'max_depth': 5},
    {'n_estimators': 150, 'learning_rate': 0.05, 'max_depth': 5},
]

# TRAIN + EVALUATE
local_gb_xgb_results = []

def evaluate(model, name, params):
    start = time.time()
    model.fit(X_train, y_train)
    elapsed = time.time() - start
    preds = model.predict(X_test)

    local_gb_xgb_results.append({
        "Phase": "Phase 1",
        "Model": name,
        "Task": "Maturity Classification",
        "Fusion": "Image Only",
        "Type": "ResNet Features",
        "Parameters": str(params),
        "Accuracy": accuracy_score(y_test, preds),
        "Precision": precision_score(y_test, preds, average="weighted"),
        "Recall": recall_score(y_test, preds, average="weighted"),
        "F1": f1_score(y_test, preds, average="weighted"),
        "Training_Time": elapsed
    })

for p in gb_params:
    evaluate(GradientBoostingClassifier(**p, random_state=42), "Gradient Boosting", p)

for p in xgb_params:
    evaluate(XGBClassifier(**p, random_state=42, eval_metric="logloss"), "XGBoost", p)

results.extend(local_gb_xgb_results)

results_df = pd.DataFrame(local_gb_xgb_results).sort_values("Accuracy", ascending=False)
print("\n=== MODEL COMPARISON ===")
print(results_df.round(4).to_string(index=False))
```

# Naiye Bayes And Random Forest

## 📦 Imports & Paths

```python
import os, cv2, warnings, time
import numpy as np
import pandas as pd

from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_decomposition import CCA
from sklearn.linear_model import LogisticRegression
from sklearn.base import clone
from tensorflow.keras.preprocessing.image import ImageDataGenerator

warnings.filterwarnings("ignore")

IMAGE_DIR  = "/content/images"
EXCEL_PATH = "/content/Maturity_dataset.xlsx"

print("✅ All libraries loaded successfully")
```

## 📂 Load Dataset

```python
df = pd.read_excel(EXCEL_PATH)

# Extract numeric Fruit ID — used to prevent same fruit in train & test
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

print(f"✅ Dataset loaded: {len(df)} rows")
print(f"   Columns: {list(df.columns)}")
df.head()
```

## 🖼️ Image Feature Extraction Function

```python
def extract_image_features(img):
    img      = cv2.resize(img, (224, 224))

    # RGB colour statistics
    rgb_mean = img.mean(axis=(0, 1))
    rgb_std  = img.std(axis=(0, 1))

    # HSV colour statistics
    hsv      = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv_mean = hsv.mean(axis=(0, 1))
    hsv_std  = hsv.std(axis=(0, 1))

    # Hue histogram (16 bins)
    hue         = hsv[:, :, 0]
    hue_hist, _ = np.histogram(hue, bins=16, range=(0, 180))
    hue_hist    = hue_hist / hue_hist.sum()

    # Yellow / Green ratio — key ripeness indicator for Harumanis mango
    yellow = ((hue >= 20) & (hue <= 35)).sum() / hue.size
    green  = ((hue >= 35) & (hue <= 85)).sum() / hue.size

    # GLCM texture features
    gray   = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_r = (gray / 4).astype(np.uint8)
    glcm   = graycomatrix(gray_r, [1],
                          [0, np.pi/4, np.pi/2, 3*np.pi/4],
                          levels=64, symmetric=True, normed=True)
    texture = [graycoprops(glcm, p).mean()
               for p in ["contrast","homogeneity","energy","correlation","dissimilarity"]]

    # Edge density
    edges        = cv2.Canny(gray, 50, 150)
    edge_density = edges.sum() / edges.size

    return np.hstack([rgb_mean, rgb_std, hsv_mean, hsv_std,
                      [yellow, green], hue_hist, texture, [edge_density]])

print("✅ Feature extractor defined")
```

## 🏗️ Build Dataset (Two-Sided Image Average + Weight)

```python
import os
print(os.listdir("images")[:20])
```

```python
imgs, weights, y_rip, y_grd, groups = [], [], [], [], []

for fid, g in df.groupby("Fruit_ID"):
    if len(g) != 2:
        continue

    r1, r2 = g.iloc[0], g.iloc[1]

    img1 = cv2.imread(os.path.join(IMAGE_DIR, r1["Fruit No"]))
    img2 = cv2.imread(os.path.join(IMAGE_DIR, r2["Fruit No"]))

    if img1 is None or img2 is None:
        print(f"⚠️  Could not load images for Fruit ID {fid}. Skipping.")
        continue

    img1 = cv2.resize(img1, (224, 224))
    img2 = cv2.resize(img2, (224, 224))

    imgs.append((img1 + img2) / 2)
    # weights.append([r1["Mass(kg)"]]) # Commented out as 'Mass(kg)' column does not exist
    y_rip.append(r1["Maturity"])
    y_grd.append(r1["Maturity"])
    groups.append(fid)

imgs    = np.array(imgs, dtype=np.float32)
weights = np.array(weights) # This will be empty, or needs to be handled differently if weights are still desired.
groups  = np.array(groups)

print(f"✅ Dataset built: {len(imgs)} fruit samples")
```

## 🔤 Encode Labels

```python
rip_enc = LabelEncoder()
grd_enc = LabelEncoder()

y_rip = rip_enc.fit_transform(y_rip)
y_grd = grd_enc.fit_transform(y_grd)

print(f"✅ Ripeness classes : {list(rip_enc.classes_)}")
print(f"✅ Grade classes    : {list(grd_enc.classes_)}")
```

## ✂️ Group-Aware Train/Test Split

```python
# GroupShuffleSplit — same fruit NEVER appears in both train and test
gss = GroupShuffleSplit(test_size=0.25, random_state=42)
train_idx, test_idx = next(gss.split(imgs, y_rip, groups))

imgs_tr, imgs_te = imgs[train_idx], imgs[test_idx]
# wt_tr,   wt_te   = weights[train_idx], weights[test_idx] # Commented out as 'weights' is empty
yr_tr,   yr_te   = y_rip[train_idx],   y_rip[test_idx]
yg_tr,   yg_te   = y_grd[train_idx],   y_grd[test_idx]

print(f"✅ Train: {len(imgs_tr)} fruits | Test: {len(imgs_te)} fruits")
```

## 🔄 Data Augmentation (Train Set Only)

```python
augmenter = ImageDataGenerator(
    rotation_range=15,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2]
)

aug_imgs, aug_yr, aug_yg = [], [], []

for img, yr, yg in zip(imgs_tr, yr_tr, yg_tr):
    img_r    = img.reshape(1, *img.shape)
    aug_iter = augmenter.flow(img_r, batch_size=1)
    for _ in range(2):      # 2 augmented copies per original image
        aug_imgs.append(next(aug_iter)[0])
        aug_yr.append(yr)
        aug_yg.append(yg)

imgs_tr_aug = np.vstack([imgs_tr, np.array(aug_imgs)])
yr_tr_aug   = np.hstack([yr_tr,   np.array(aug_yr)])
yg_tr_aug   = np.hstack([yg_tr,   np.array(aug_yg)])

print(f"✅ After augmentation — Train size: {len(imgs_tr_aug)}")
```

## ⚙️ Feature Extraction & Scaling

```python
print("⏳ Extracting features from training images...")
Xi_tr = np.array([extract_image_features(i.astype(np.uint8)) for i in imgs_tr_aug])

print("⏳ Extracting features from test images...")
Xi_te = np.array([extract_image_features(i.astype(np.uint8)) for i in imgs_te])
print(f"✅ Feature vector size: {Xi_tr.shape[1]}")

# RobustScaler — robust to outliers in image feature distributions
sc_img = RobustScaler()

Xi_tr     = sc_img.fit_transform(Xi_tr)
Xi_te     = sc_img.transform(Xi_te)

# Combined feature matrix (image only) — used in Phase 1 & Early Fusion
X_tr = Xi_tr
X_te = Xi_te

# Early Fusion dataset — same as combined, kept separately for clarity
X_early_tr = Xi_tr
X_early_te = Xi_te

print(f"✅ Scaling done")
print(f"   Combined feature matrix — Train: {X_tr.shape} | Test: {X_te.shape}")
```

### 🔧 Parameter Sets (3 Variants Each)

```python
# Naive Bayes — var_smoothing controls numerical stability
nb_params = [
    {'var_smoothing': 1e-9},    # Default — minimal smoothing
    {'var_smoothing': 1e-7},    # Moderate smoothing
    {'var_smoothing': 1e-5},    # High smoothing
]

# Random Forest — trees, depth, split threshold
rf_params = [
    {'n_estimators': 50,  'max_depth': 3,    'min_samples_split': 2},   # Shallow
    {'n_estimators': 100, 'max_depth': 5,    'min_samples_split': 5},   # Balanced
    {'n_estimators': 150, 'max_depth': None, 'min_samples_split': 2},   # Full-depth
]

print("✅ Parameter sets defined")
```

### 🏋️ Train & Evaluate — Phase 1

```python
local_nb_rf_results = []

def evaluate_p1(model, name, params):
    start = time.time()
    model.fit(X_tr, yr_tr_aug)
    elapsed = time.time() - start
    preds = model.predict(X_te)
    local_nb_rf_results.append({
        "Phase": "Phase 1",
        "Model": name,
        "Task": "Maturity Classification",
        "Fusion": "Image Only",
        "Type": "Augmented Features",
        "Parameters": str(params),
        "Accuracy": accuracy_score(yr_te, preds),
        "Precision": precision_score(yr_te, preds, average="weighted", zero_division=0),
        "Recall": recall_score(yr_te, preds, average="weighted", zero_division=0),
        "F1": f1_score(yr_te, preds, average="weighted", zero_division=0),
        "Training_Time": round(elapsed, 4)
    })

# Naive Bayes — 3 variants
print("🔵 Training Naive Bayes...")
for p in nb_params:
    evaluate_p1(GaussianNB(**p), "Naive Bayes", p)
    print(f"   ✔ var_smoothing = {p['var_smoothing']}")

# Random Forest — 3 variants
print("\n🟢 Training Random Forest...")
for p in rf_params:
    evaluate_p1(RandomForestClassifier(**p, random_state=42), "Random Forest", p)
    print(f"   ✔ n_estimators={p['n_estimators']}, max_depth={p['max_depth']}")

results.extend(local_nb_rf_results)
print("\n✅ Phase 1 training complete and results added to global list!")
```

# KNN and Decision Tree Training on Image Dataset



```python
import os
import numpy as np
import pandas as pd
from PIL import Image

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image

IMAGE_FOLDER = "/content/images"
EXCEL_PATH = "/content/Maturity_dataset.xlsx"

df = pd.read_excel(EXCEL_PATH)

# Extract numeric Fruit ID — used to prevent same fruit in train & test
df["Fruit_ID"] = df["Fruit No"].str.extract(r"(\d+)")

print("Columns:", df.columns.tolist())

# -------- Image Preprocessing --------
def load_image(path, size=(64, 64)):
    img = Image.open(path).convert("RGB")
    img = img.resize(size)
    return np.array(img).flatten()

X_img = []
X_w = []
y_ripeness = []
y_grade = []
groups = []

for _, row in df.iterrows():
    img_name = str(row["Fruit No"]).strip()
    img_path = os.path.join(IMAGE_FOLDER, img_name)

    if os.path.exists(img_path):
        X_img.append(load_image(img_path))
        X_w.append([0])
        y_ripeness.append(row["Maturity"])
        y_grade.append("Unknown")
        group_id = "".join(filter(str.isdigit, img_name))
        groups.append(group_id)
    else:
        print("Missing image:", img_path)

X_img = np.array(X_img)
X_w = np.array(X_w)
y_ripeness = np.array(y_ripeness)
y_grade = np.array(y_grade)
groups = np.array(groups)

print("Image feature shape:", X_img.shape)
print("Weight feature shape:", X_w.shape)
```

```python

# Image preprocessing function
def load_image(path, size=(64, 64)):
    img = Image.open(path).convert("RGB")
    img = img.resize(size)
    return np.array(img).flatten()
```

```python
from sklearn.preprocessing import LabelEncoder

le_ripeness = LabelEncoder()
le_grade = LabelEncoder()

y_ripeness = le_ripeness.fit_transform(y_ripeness)
y_grade = le_grade.fit_transform(y_grade)
```

```python
from sklearn.model_selection import GroupShuffleSplit

gss = GroupShuffleSplit(test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(X_img, y_ripeness, groups))

# Image
X_img_train, X_img_test = X_img[train_idx], X_img[test_idx]

# Weight
X_w_train, X_w_test = X_w[train_idx], X_w[test_idx]

# Labels
yr_train, yr_test = y_ripeness[train_idx], y_ripeness[test_idx]
yg_train, yg_test = y_grade[train_idx], y_grade[test_idx]

print("Train size:", len(X_img_train))
print("Test size:", len(X_img_test))
```

```python
local_knn_dt_results = []

# -------- KNN --------
k_values = [3, 5, 7, 9]

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_img_train, yr_train)
    y_pred = model.predict(X_img_test)

    local_knn_dt_results.append({
        "Phase": "Phase 1",
        "Model": "KNN",
        "Task": "Maturity Classification",
        "Fusion": "Image Only",
        "Type": "Raw Pixels",
        "Modification": f"k = {k}",
        "Accuracy": accuracy_score(yr_test, y_pred),
        "Precision": precision_score(yr_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(yr_test, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(yr_test, y_pred, average="weighted", zero_division=0)
    })


# -------- DECISION TREE --------
depths = [3, 5, 10, None]

for d in depths:
    model = DecisionTreeClassifier(max_depth=d, random_state=42)
    model.fit(X_img_train, yr_train)
    y_pred = model.predict(X_img_test)

    local_knn_dt_results.append({
        "Phase": "Phase 1",
        "Model": "Decision Tree",
        "Task": "Maturity Classification",
        "Fusion": "Image Only",
        "Type": "Raw Pixels",
        "Modification": f"max_depth = {d}",
        "Accuracy": accuracy_score(yr_test, y_pred),
        "Precision": precision_score(yr_test, y_pred, average="weighted", zero_division=0),
        "Recall": recall_score(yr_test, y_pred, average="weighted", zero_division=0),
        "F1": f1_score(yr_test, y_pred, average="weighted", zero_division=0)
    })

results.extend(local_knn_dt_results)
print(f"✅ KNN & Decision Tree metrics added. Total results in list: {len(results)}")
```

```python

results_df = pd.DataFrame(results)

results_df[["Accuracy", "Precision", "F1", "Recall"]] = (
    results_df[["Accuracy", "Precision", "F1", "Recall"]].round(4)
)

results_df
```
